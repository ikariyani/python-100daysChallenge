# PANDAS - Import Data | CSV, Excel, JSON

```python
import pandas as pd
```

## 1. Import Data CSV

```python
# Pada Jupyter Notebook, file yang akan digunakan pada Pandas harus berada pada folder yang sama dengan file ini.

file = 'Internet Speed 2022.csv'

csv = pd.read_csv(file)

csv
```

## 2. Import Data Excel

```python
# Untuk mengubah file CSV ke file Excel, perlu untuk download module "openpyxl" terlebih dahulu

!pip install openpyxl

```

```python
# Setelah menginstal module "openpyxl" maka file dapat disimpan dalam bentuk Excel.


csv.to_excel('Internet_Speed_2022_Excel.xlsx')
```

```python
# Setelah file tersimpan dalam bentuk Excel, maka file excel sudah dapat dibaca.
# Pada Pandas, file excel yang sudah ada dapat langsung dibaca

excel = pd.read_excel('Internet_Speed_2022_Excel.xlsx')

excel
```

## 3. Import Data JSON

```python
# Convert file csv yang sudah ada ke file "json" terlebih dahulu 

csv.to_json('Internet_Speed_2022_Excel.json')
```

```python
# Apabila file sudah tersimpan dalam bentuk json, maka file sudah dapat dibaca.
# Pada Pandas, file json yang sudah ada dapat langsung dibaca.

json = pd.read_json('Internet_Speed_2022_Excel.json')


json
```
