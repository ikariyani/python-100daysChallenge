# PANDAS - Data Frame | Data Slicing

```python
import pandas as pd
```

## 1. Standard Slicing ([])

### A. Slicing Single Column

```python
# Standard Slicing bisa digunakan ketika kita ingin mengakses satu kolom tertentu

file = 'Marvel Movies.csv'
df = pd.read_csv(file)

df
```

```python
# Berikut contoh jika ingin mengambil semua data pada satu kolom yaitu "name"

df['name']
```

```python
df['ID']
```

### B. Slicing Multiple Columns

```python
# Standard Slicing juga bisa digunakan ketika kita ingin mengakses kolom tertentu lebih dari satu

df[['name', 'ID']]
```

```python
df[['name', 'ALIGN']]
```

### C. Slicing Object

```python
# Standard Slicing bisa digunakan untuk memilih jumlah data/objek tertentu

df[0:10]
```

```python
# Standar Slicing bisa digunakan untuk mengambil "row" tertentu
# Rumus : df[row ke -: hingga row ke-: lompat -angka]

df[1:10:2]
```

### D. Selecting Multiple Rows and Single Column

```python
# Standard Slicing juga dapat digunakan untuk memilih beberapan rows dalam satu kolom

df[0:10]['name']
```

```python
df[10:20][['name', 'EYE']]
```

## 2. loc

```python
# .loc berfungsi untuk memilih baris dan kolom berdasarkan labelnya
```

### A. .loc Single Row

```python
# Rumus : df.loc[ baris ke- ]

df.loc[3]
```

```python
df.loc[100]
```

### B. .loc Multiple Rows

```python
# .loc bisa digunakan untuk memilih baris tertentu
# Rumus : df.loc[ baris ke- , baris ke- ]


df.loc[[7, 30]]
```

```python
df.loc[[4, 6, 9, 12]]
```

### C. .loc Single Row and Single Column

```python
# .loc bisa digunakan untuk memilih baris dan kolom tertentu
# Rumus : df.loc[ baris ke-, nama kolom ]

df.loc[3, 'name']
```

```python
df.loc[10, 'name']
```

### D. .loc Single Row and Multiple Columns

```python
# .loc bisa digunakan untuk memilih kolom lebih dari 1

df.loc[[4], ['name', 'ALIGN', 'EYE']]
```

```python
df.loc[[8], ['name', 'ID', 'ALIGN', 'HAIR']]
```

### E. Slicing Object with .loc

```python
# Rumus : df.loc[ baris ke- : baris ke - ]

df.loc[0:10]
```

```python
# Rumus : df.loc[ baris ke - : baris ke- : lompat -angka]

df.loc[1:10:2]
```

## 3. .iloc

```python
# .iloc berfungsi untuk memilih baris dan kolom berdasarkan letak value nya
```

### A. .iloc Single Row

```python
df.iloc[3]
```

```python
df.iloc[150]
```

### B. .iloc Multiple Rows

```python
df.iloc[[10, 20]]
```

```python
df.iloc[[5, 10, 15, 20, 25]]
```

### C. .iloc Single Row and Multiple Columns

```python
df.iloc[[50], [0,1,2]]
```

```python
df.iloc[[100], [0,1,2,3,4]]
```

### D. .iloc Multiple Rows and Single Column

```python
df.iloc[[1,2,3,4,5,], [0]]
```

```python
df.iloc[[10, 20, 30, 40, 50], [0]]
```

### E. .iloc Multiple Rows and Multiple Columns

```python
df.iloc[[10, 20, 30, 40], [0,1]]
```

```python
df.iloc[[5, 10, 15, 20, 25, 30], [0,1,2,3,4,5]]
```

### F. Slicing Object with .iloc

```python
df.iloc[1:10:2]
```

```python
df.iloc[1:50:5]
```
