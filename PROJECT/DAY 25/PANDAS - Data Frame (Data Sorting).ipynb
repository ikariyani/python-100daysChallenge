# PANDAS - Data Frame | Data Sorting

Sorting pada Pandas digunakan untuk mengurutkan data pada sebuah dataframe.

```python
import pandas as pd
```

```python
file = "Penyebab Kematian di Indonesia yang Dilaporkan - Raw.csv"
```

```python
df = pd.read_csv(file)

df
```

## 1. Sort by Single Column

```python
# Untuk mengurutkan data yang dibutuhkan, maka dapat menggunakan fungsi "sort_values" 
# Jika ingin mengurutkan data berdasarkan 1 kolom saja, maka cukup label kolom tersebut yang dicantumkan pada fungsi


df.sort_values(by = ['Cause'])
```

```python
df.sort_values(['Total Deaths'])
```

## 2. Sort by Multiple Columns

```python
# Jika ingin mengurutkan data pada beberapa kolom, maka label kolom tersebut bisa dituliskan semua pada fungsi meskipun lebih dari 1 kolom


df.sort_values(['Cause', 'Total Deaths'])
```

```python
# Apabila ingin menampilkan data berdasarkan baris tertentu, maka bisa menggunakan "head( jumlah baris )"


df.sort_values(['Total Deaths','Cause', 'Year']).head(150)
```

## 3. Sort by Ascending and Descending

### A. Ascending

```python
# Apabila mengurutkan data menggunakan "ascending = True", maka data akan terurutkan dari data yang terkecil, apabila data berupa text maka akan diurutkan berdasarkan urutan huruf abjad


df.sort_values(['Year', 'Type'], ascending = True)
```

```python
df.sort_values(['Cause', 'Year'], ascending = True).head(10)
```

### B. Descending

```python
# Apabila argumen "ascending" = "False", maka akan mengurutkan secara Descending
# Descending memiliki arti bahwa data akan diurutkan berkebalikan dengan "ascending = True"


df.sort_values(['Total Deaths','Cause'], ascending = False)
```

```python
df.sort_values(['Total Deaths', 'Year'], ascending = False).head(10)
```
