# PANDAS - Data Frame | Data Restructuring

```python
import pandas as pd
```

```python
file = "KSN_SMA_2008-2021.xlsx"

df = pd.read_excel(file)

df
```

## 1. PIVOT

"Pivot Table" digunakan untuk mengubah bentuk tabel dalam data frame seperti yang diinginkan

```python
df.pivot_table(index = 'province', columns = 'subject', values = 'medal', aggfunc = 'first')
```

```python
df.pivot_table(index = 'year', columns = 'province', values = 'medal', aggfunc = 'count')
```

## 2. Stack

Secara sederhana method stack akan melakukan pivoting kolom menjadi index.

```python
df.stack()
```

```python
df.stack().to_frame()
```

## 3. Unstack

Fungsi "Unstack" akan mengembalikan index menjadi kolom

```python
df.stack().unstack()
```
