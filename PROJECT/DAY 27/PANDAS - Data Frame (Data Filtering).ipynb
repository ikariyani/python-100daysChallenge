# PANDAS - Data Frame | Data Filtering

```python
import pandas as pd
```

```python
file = "Mineral Ores Around The World Cleaned.csv"

df = pd.read_csv(file)


df
```

```python
df.columns
```

## 1. Boolean Mask

Boolean Masking dapat digunakan untuk memfilter data berdasarkan pada kondisi tertentu

```python
df.dev_stat 
```

```python
df.dev_stat == 'producer'
```

```python
df[df.dev_stat == 'Producer']
```

```python
df[df.country == 'Indonesia'][df.dev_stat == 'Producer']
```

## 2. Loc

Loc merupakan metode yang dapat digunakan jika kita ingin mengambil kolom tertentu Loc dapat digunakan untuk mengakses sekelompok baris dan kolom dengan label.

```python
df
```

```python
df.loc[10]
```

```python
df.loc[1:100]
```

```python
df.loc[[1, 2, 3, 4, 5], ['state', 'oper_type', 'ore', 'prod_size', 'dev_stat']]
```

```python
df.loc[df['country'].str.contains('Indonesia')]
```

## 3. Iloc

Iloc digunakan untuk memfilter data berdasarkan index

```python
df
```

```python
df.iloc[50:100]
```

```python
df.iloc[:10, [0, 5, 11, 13, 15]]
```

## 4. Query

```python
df
```

```python
df.query("oper_type == 'Surface'")
```

```python
df.query("dev_stat == 'Producer' & work_type == 'Underground'")
```
