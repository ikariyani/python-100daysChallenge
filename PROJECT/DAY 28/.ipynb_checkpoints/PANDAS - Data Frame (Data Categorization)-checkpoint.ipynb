```python
import pandas as pd
```

```python
file = "daily_weather_data.csv"

df = pd.read_csv(file)

df
```

## 1. Metode 1

Pengkategorian data dapat menggunakan beberapa metode. Metode yang pertama adalah dengan membuat tambahan kolom untuk kategori data. Setelah itu, setiap value akan dikelompokkan sesuai dengan kategorisasi yang dibuat.

Klasifikasi angin berdasarkan kecepatan menurut Beaufort :

|Angin||Kecepatan (km/jam)||Klasifikasi||Fenomena|
|:---:||:---:||:---:||:---:|
|0||<1.6||Calm||Asap bergerak vertikal|
|1||1.6 - 4.8||Light Air||Arah angin ditunjukkan oleh asap|
|2||6.4 - 11.3||Light Breeze||Terasa pada muka, daun berisik, penunjuk angin mulai bergerak|
|3||12.9 - 19.3||Gentle Breeze||Daun dan ranting bergerak tetap, bendera berkibar|
|4||20.9 - 29.0||Moderate Breeze||Menggerakkan tangkai kecil, menghembus debu dan kertas|
|5||30.6 - 38.6||Fresh Breeze||Pohon kecil yang berdaun mulai bergoyang|
|6||40.2 - 49.9||Strong Breeze||Dahan besar begoyang, kawat telpon berdering|
|7||51.5 - 61.1||Moderate Gale||Seluruh pohon bergoyang|
|8||62.8 - 74.0||Fresh Gale||Ranting pohon patah|
|9||75.6 - 86.9||Strong Gale||Bangunan ringan rusak|
|10||88.5 - 101.4||Whole Gale||Bangunan rusak, pohon tumbang|
|11||103.0 - 120.7||Storm||Kerusakan bangunan sangat banyak|
|12||>120.7||Hurricane||Kerusakan besar, dahsyat dan meluas|

```python
df
```

```python
# 1. Buatlah kolom tambahan untuk meletakkan kategorisasi data yang diisi oleh value yang belum ditentukan


df['category'] ='undefined'

df
```

```python
# 2. Setelah kolom kategori tersebut dibuat, maka data harus dikelompokkan sesuai yang diinginkan satu persatu pada kolom "category"

df.loc[df['wspd'] <1.6, 'category' ] = 'Calm'

df[df['category'] == 'Calm']
```

```python
df.loc[df['wspd'] >1.6 and  4.8, 'category' ] = 'Light Air'

df[df['category'] == 'Light Air']
```
