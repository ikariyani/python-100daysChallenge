## 2. Metode 2

Pengkategorian data menggunakan metode yang ke dua adalah dengan menggunakan fungsi dan logika if-else.

```python
df
```

```python
def klasifikasi(x) :
    if x <1 :
        return 'Calm'
    elif (x >=1 and x <6) :
        return 'Light Air'
    elif (x >=6 and x <12) :
        return 'Light Breeze'
    elif (x >=12 and x <20) :
        return 'Gentle Breeze'
    elif (x >=20 and x <29) :
        return 'Moderate Breeze'
    elif (x >=29 and x <39) :
        return 'Fresh Breeze'
    elif (x >=39 and x <50) :
        return 'Strong Breeze'
    elif (x >=50 and x<62) :
        return 'Near Gale'
    elif (x >=62 and x <75) :
        return 'Gale'
    elif (x >=75 and x <89) :
        return 'Strong Gale'
    elif (x >=89 and x <103) :
        return 'Storm'
    elif (x >=103 and x <117) :
        return 'Violent Storm'
    elif (x >=117) :
        return 'Hurricane'
    else :
        return 'Undefined'

df['category metode ke-2'] = df['wspd'].apply(klasifikasi)
```

## 3. Metode 3

Metode ke -3 untuk mengklasifikasikan data menggunakan fungsi "cut()". Dalam penggunaan fungsi ini, batas antar klasifikasi harus sangat jelas.

```python
# Buatlah batasan klasifikasi dalam bentuk array, pada contoh berikut berupa "bins"
# Buatlah nama dari setiap klasifikasi, pada contoh berikut berupa "kelas"

bins = [0, 0.99, 5.99, 11.99, 19.99, 28.99, 37.99, 49.99, 61.99, 74.99, 88.99, 102.99, 117.99, 1000]
kelas = ['Calm', 'Light Air', 'Light Breeze', 'Gentle Breeze', 'Moderate Breeze', 'Fresh Breeze', 
         'Strong Breeze', 'Near Gale', 'Gale', 'Strong Gale', 'Storm', 'Violent Storm', 'Hurricane']
```

```python
# Setelah batasan klasifikasi dan nama klasifikasi dibuat, maka dapat menggunakan fungsi berikut untuk pengklasifikasian menggunakan fungsi "cut()"
# Rumus : df[kolom kategori] = pd.cut( df. {kolom yang akan dikategorikan} , {array rentang klasifikasi} , labels = {nama setiap klasifikasi} )


df['category metode ke-3'] = pd.cut(df.wspd, bins, labels = kelas)
```
