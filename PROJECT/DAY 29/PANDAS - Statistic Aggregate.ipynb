# PANDAS - Statistic Aggregate

```python
import pandas as pd
```

```python
file = "hurricane_clean_dataset.csv"

df = pd.read_csv(file)

df
```

## 1. Fungsi Agregasi

Mengolah suatu data dapat menggunakan function atau fungsi .agg() untuk mengaplikasikan analisis statistik deskriptif.

```python
# Total kematian akibat dari badai dapat diketahui dengan menggunakan fungsi "agg( 'sum' )"


df['Deaths'].agg('sum')
```

```python
# Jumlah badai yang terjadi dapat diketahui dengan menggunakan fungsi "agg( 'count' )"


df['Name'].agg('count')
```

```python
# Rata - rata kecepatan angin pada badai yang terjadi dapat diketahui menggunakan fungsi "agg( 'mean' )"

df['Wind Speed (mph)'].agg('mean')
```

```python
# Kerugian dan korban kematian minimum dari badai yang terjadi dapat diketahui dengan menggunakan fungsi "agg( 'min' )"


df[['Damages ($)', 'Deaths']].agg('min')
```

```python
# Kerugian dan korban kematian maksimum dari badai yang terjadi dapat diketahui dengan menggunakan fungsi "agg( 'max' )"


df[['Damages ($)', 'Deaths']].agg('max')
```

## 2. GroupBy

Groupby adalah fungsi untuk mengelompokkan data lebih dari satu kolom berdasarkan satu kolom tertentu.

```python
df
```

```python
df.groupby(['Year'])['Deaths'].sum()
```

```python
df.groupby(['Year']).sum().sort_values('Deaths', ascending = False)
```

```python
df.groupby(['Year'])['Name'].count()
```

```python
df.groupby(['Year'])['Wind Speed (mph)'].mean()
```

```python
df.groupby(['Year']).min().sort_values('Year', ascending = False)
```

```python
df.groupby(['Year']).min().sort_values('Deaths', ascending = False)
```
