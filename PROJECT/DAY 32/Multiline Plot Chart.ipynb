```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('HDI.csv')

df
```

```python
asean = df.loc[((df['Country'] == 'Indonesia') | 
        (df['Country'] == 'Malaysia') | 
        (df['Country'] == 'Singapore') | 
        (df['Country'] == 'Philippines') |
        (df['Country'] == 'Thailand') |
        (df['Country'] == 'Brunei Darussalam') |
        (df['Country'] == 'Cambodia') |
        (df['Country'] == "Lao People's Democratic Republic") |
        (df['Country'] == 'Myanmar') |
        (df['Country'] == 'Viet Nam') |
        (df['Country'] == 'Timor Leste')
       )]

asean
```

```python
asean2 = asean.drop(columns= ['HDI Rank'])

asean2
```

```python
asean3 = asean2.set_index('Country')

asean3
```

```python
asean4 = asean3.astype(np.double).transpose()

asean4
```

```python
asean4.plot(kind='line')

plt.title('Hello World')
plt.xlabel('Hello')
plt.ylabel('World')

plt.show()
```

## All Dataset

```python
country_all = df.drop(columns=['HDI Rank']).set_index('Country')

country_all
```

```python
country10 = country_all.head(10).to_numpy().astype(np.str_)

country10
```

```python
country10_str = np.char.replace(country10, '..', 'NaN')

country10_str
```

```python
country10_clean = country10_str.astype(np.double)

country10_clean
```

```python
variabel = df['Country'].head(10).to_numpy()

variabel
```

```python
for idx, x in enumerate(country10_clean):
    plt.plot(tahun, x, label= variabel[idx])
    
plt.title('hello world')

plt.legend()
    
plt.show()
```
