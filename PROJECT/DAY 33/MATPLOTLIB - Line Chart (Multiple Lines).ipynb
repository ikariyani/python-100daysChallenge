# MATPLOTLIB - Line Chart | Multiple Lines

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
file = "HDI.csv"

df = pd.read_csv(file)

df
```

```python
def kategori (x) :
    if (x == 'Indonesia' or x == 'Malaysia' or x == 'Singapore' or x == 'Philippines' or x == 'Thailand' or x == 'Brunei Darussalam' or x == 'Cambodia' or x == "Lao People's Democratic Republic" or x == 'Myanmar' or x == 'Viet Nam') :
         return 'Asia Tenggara'
    
    else :
        return 'Non Asia Tenggara'
    


df['Kategori'] = df['Country'].apply(kategori)
```

```python
df
```

```python
Asean =  df[df['Kategori'] == 'Asia Tenggara']
```

```python
Asean
```

```python
Asean = Asean.reset_index()

Asean
```

```python
Asean = Asean.drop(columns = ['index', 'HDI Rank', 'Kategori'])
```

```python
Asean
```

```python
Asean = Asean.set_index('Country')

Asean
```

```python
Asean = Asean.to_numpy().astype(np.double)

Asean
```

```python
Tahun = np.arange(1990, 2020, 1)

Tahun 
```

```python
ASEAN = df[df['Kategori'] == 'Asia Tenggara']

Negara = ASEAN['Country'].to_numpy()

Negara
```

```python
import matplotlib.font_manager as fm
```

```python
path = "MADEAvenuePERSONALUSE-Regular.otf"
path2 = "Philosopher-Regular.ttf"
path3 = "Quicksand_Bold.otf"


font1 = fm.FontProperties(fname = path, size = 15)
font2 = fm.FontProperties(fname = path2, size = 15)
font3 = fm.FontProperties(fname = path3, size = 10)
font4 = fm.FontProperties(fname = path4, size = 10)
font5 = fm.FontProperties(fname = path2, size = 10)
```

```python
plt.figure(figsize=(10, 5))

for idx, x in enumerate(Asean):
    plt.plot(Tahun, x, label = Negara[idx], marker = '.')


plt.title("Indeks Pembangunan Manusia Negara Anggota ASEAN", fontproperties = font1)
plt.xlabel("Tahun", fontproperties = font2)
plt.ylabel("IPM", fontproperties = font2)

plt.xticks(np.arange(1990, 2021, 2),fontproperties = font3)
plt.yticks(fontproperties = font3)

plt.legend(bbox_to_anchor = (1.02, 1), labelspacing = 0.8, prop = font5)

plt.grid(linestyle = 'dashed')

plt.show()
```
