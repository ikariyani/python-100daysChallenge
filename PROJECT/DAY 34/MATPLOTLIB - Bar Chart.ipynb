# MATPLOTLIB - Bar Chart 

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
file = "HDI.csv"

df = pd.read_csv(file)

df
```

## Fetch data HDI Indonesia

```python
indo_HDI = df[df['Country'] == 'Indonesia']
```

```python
indo_HDI
```

```python
drop = indo_HDI.drop(columns = ['HDI Rank', 'Country'])

drop
```

```python
drop.reset_index()
```

```python
indo_HDI = drop.to_numpy()

indo_HDI
```

```python
indo_HDI = indo_HDI.astype(float)

indo_HDI.ndim
```

```python
tahun = np.arange(1990, 2020, 1)

tahun.ndim
```

```python
indo_HDI = indo_HDI.reshape(30,)

indo_HDI
```

```python
import matplotlib.font_manager as fm
```

```python
path = "MADEAvenuePERSONALUSE-Regular.otf"
path2 = "Philosopher-Regular.ttf"
path3 = "Quicksand_Bold.otf"


font1 = fm.FontProperties(fname = path, size = 20)
font2 = fm.FontProperties(fname = path2, size = 15)
font3 = fm.FontProperties(fname = path3, size = 10)
font4 = fm.FontProperties(fname = path2, size = 9)
font5 = fm.FontProperties(fname = path2, size = 6)
font6 = fm.FontProperties(fname = path2, size = 8)
```

```python
plt.figure(figsize = (8,4))

bar = plt.bar(tahun, indo_HDI, label = 'IPM', color = '#25735d')

plt.bar_label(bar, rotation = 90, padding = 5, fontproperties = font4)

plt.title('Indeks Pembangunan Manusia Negara Indonesia', fontproperties = font1)
plt.xlabel('Tahun',fontproperties = font2 )
plt.ylabel('IPM', fontproperties = font2)

plt.xticks(np.arange(1990, 2021, 2), fontproperties = font3)
plt.yticks(np.arange(0, 0.95, 0.1), fontproperties = font3)

plt.legend(prop = font6)
plt.grid(axis = 'y', linestyle = 'dashed')


plt.show()
```

## Fetch Data HDI ASEAN

```python
df
```

```python
def kategori (x) :
    if (x == 'Indonesia' or x == 'Malaysia' or x == 'Singapore' or x == 'Philippines' or x == 'Thailand' or x == 'Brunei Darussalam' or x == 'Cambodia' or x == "Lao People's Democratic Republic" or x == 'Myanmar' or x == 'Viet Nam') :
         return 'Asia Tenggara'
    
    else :
        return 'Non Asia Tenggara'
    


df['Kategori'] = df['Country'].apply(kategori)
```

```python
Asean =  df[df['Kategori'] == 'Asia Tenggara']

Asean
```

```python
Asean = Asean.drop(columns = ['HDI Rank', 'Kategori'])

Asean
```

```python
Asean_rank = Asean[['2019', 'Country']].sort_values('2019', ascending = True)

Asean_rank
```

```python
tahun2019 = Asean_rank['2019'].to_numpy()

tahun2019
```

```python
Negara = Asean_rank['Country'].to_numpy()

Negara
```

```python
path = "MADEAvenuePERSONALUSE-Regular.otf"
path2 = "Philosopher-Regular.ttf"
path3 = "Quicksand_Bold.otf"


font1 = fm.FontProperties(fname = path, size = 17)
font2 = fm.FontProperties(fname = path2, size = 15)
font3 = fm.FontProperties(fname = path2, size = 8)
font4 = fm.FontProperties(fname = path2, size = 12)
```

```python
plt.figure(figsize = (8.5,4))

bars = plt.barh(Negara, tahun2019, 
         color = ['#dfeae2', '#b4d6c1', '#88deb1', '#69c7af', '#4eadaf', '#377a97', '#254a7f', '#162068', '#140850', '#0b132b'])

plt.bar_label(bars, label_type = 'edge', fontproperties = font3)


plt.title('Indeks Pembangunan Manusia Tahun 2019\nNegara Anggota ASEAN', fontproperties = font1)
plt.xlabel('IPM',fontproperties = font2 )


plt.xticks(np.arange(0, 1.1, 0.1), fontproperties = font4)
plt.yticks(fontproperties = font4)


plt.grid(axis = 'x', linestyle = 'dashed')

plt.show()
```
