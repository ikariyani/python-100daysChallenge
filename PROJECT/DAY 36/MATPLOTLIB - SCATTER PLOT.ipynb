```python
import matplotlib.font_manager as fm
```

```python
path = "Badgeworthy.ttf"
path2 = "GrebackGrotesquePERSONAL-Regular.otf"
path3 = "Wigners.otf"


font1 = fm.FontProperties(fname = path, size = 10)
font2 = fm.FontProperties(fname = path2, size = 13)
font3 = fm.FontProperties(fname = path3, size = 20)
font4 = fm.FontProperties(fname = path2, size = 8)
```

```python
plt.figure(figsize=(8, 5))

plt.scatter(X, Y, label = 2015, color = '#fde725', s = 100)
plt.scatter(X2, Y2, label = 2016, color = '#6dcd59',s = 100,  alpha = 1)
plt.scatter(X3, Y3, label = 2017, color = '#35b779',s = 100,  alpha = 0.7)
plt.scatter(X4, Y4, label = 2018, color = '#25828e',s = 100,   alpha = 0.7)
plt.scatter(X5, Y5, label = 2019, color = '#3e4a89', s = 100, alpha = 0.7)
plt.scatter(X6, Y6, label = 2020, color = '#440d54', s = 100, alpha = 0.6)

plt.xlabel("Government Trust", fontproperties = font2, labelpad = 10)
plt.ylabel("GDP Per Capita", fontproperties = font2, labelpad = 10)
plt.title("Government Trust and GDP Per Capita\nof Countries Around The World in 2015 - 2020", fontproperties = font3, pad = 25)

plt.legend(prop = font4)


plt.xticks(fontproperties = font4)
plt.yticks(fontproperties = font4)

plt.grid(linestyle = 'dashed')


plt.show()
```
