# MATPLOTLIB - Stem Plot


```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
file = "oil_consumption.csv"

df = pd.read_csv(file)

df
```

```python
X = df['year']

X
```

```python
Y = df['oil_energy_use']

Y
```

```python
import matplotlib.font_manager as fm
```

```python
path = "MADEAvenuePERSONALUSE-Regular.otf"
path2 = "Philosopher-Regular.ttf"
path3 = "Quicksand_Bold.otf"
path4 = "GrebackGrotesquePERSONAL-Regular.otf"



font1 = fm.FontProperties(fname = path, size = 23)
font2 = fm.FontProperties(fname = path2, size = 15)
font3 = fm.FontProperties(fname = path3, size = 10)
font4 = fm.FontProperties(fname = path4, size = 10)
font5 = fm.FontProperties(fname = path2, size = 10)
```

```python
plt.figure(figsize = (10, 5))

plt.stem(X, Y, label = 'Oil Consumption', linefmt = '#668b69', markerfmt = '#2d3c1f', basefmt = '#206261')


plt.xticks(np.arange(1960, 2025, 5), fontproperties = font4)
plt.yticks(np.arange(0, 1000, 100), fontproperties = font4)

plt.title("Oil Consumption in Indonesia", fontproperties = font1)
plt.xlabel('Year', fontproperties = font2)
plt.ylabel ('Oil Energy Use', fontproperties = font2)

plt.grid(linestyle = 'dashed')

plt.legend(prop = font2)

plt.show()
```
