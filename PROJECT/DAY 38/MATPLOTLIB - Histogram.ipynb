```python
import matplotlib.font_manager as fm
```

```python
path = "Amelaryas.otf"
path2 = "AUBREY1__.TTF"
path3 = "Caver Pitalsal.otf"
path4 = "Galgony.ttf"


font1 = fm.FontProperties(fname = path, size = 10)
font2 = fm.FontProperties(fname = path4, size = 13)
font3 = fm.FontProperties(fname = path3, size = 30)
font4 = fm.FontProperties(fname = path4, size = 30)
font5 = fm.FontProperties(fname = path4, size = 13)
font6 = fm.FontProperties(fname = path2, size = 10)
```

```python
plt.figure(figsize = (10,6))

plt.hist(jkt_age, label = 'Jakarta', bins = 30, alpha = 0.7, color = '#084c61')
plt.hist(bandung_age, label = 'Bandung', bins = 30, alpha = 0.7, color = '#db3a34')
plt.hist(surabaya_age, histtype = 'bar', label = 'Surabaya', bins = 30, alpha = 0.7, color = '#ffc857')


plt.title('Food Consumption in Indonesia', fontproperties = font4, pad = 20)
plt.xlabel('Years', fontproperties = font5)
plt.ylabel('Counts', fontproperties = font5)

plt.xlim(15, 70)
plt.xticks(fontproperties = font6)
plt.yticks(fontproperties = font6)


plt.legend(prop = font2)
plt.grid(linestyle = 'dashed', axis = 'y')

plt.show()
```
