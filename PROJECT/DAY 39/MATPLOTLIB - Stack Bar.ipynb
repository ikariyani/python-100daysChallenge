# MATPLOTLIB - Stack Bar

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
file ="age-between-15-and-49.csv"

df = pd.read_csv(file)

df
```

```python
def kategori (x) :
    if (x == 'Indonesia' or x == 'Malaysia' or x == 'Singapore' or x == 'Philippines' or x == 'Thailand' or x == 'Brunei' or x == 'Cambodia' or x == "Laos" or x == 'Myanmar' or x == 'Vietnam' or x == 'Timor') :
         return 'Asia Tenggara'
    
    else :
        return 'Non Asia Tenggara'

    
df['Kategori'] = df['Country'].apply(kategori)
```

```python
year_2019 = df.loc[df['Year'] == 2019]

year_2019
```

```python
Asia_Tenggara = year_2019.loc[year_2019['Kategori'] == 'Asia Tenggara']

Asia_Tenggara
```

```python
indo = Asia_tenggara.loc[Asia_tenggara['Country'] == 'Indonesia']

indo
```

```python
penyakit = indo[['Year', 'Malaria','HIV/AIDS', 'Diabetes mellitus', 'Chronic kidney disease', 'Acute hepatitis', 'Protein-energy malnutrition', 'Nutritional deficiencies']]

penyakit
```

```python
import matplotlib.font_manager as fm
```

```python
path = "MADEAvenuePERSONALUSE-Regular.otf"
path2 = "Philosopher-Regular.ttf"
path3 = "Quicksand_Bold.otf"
path4 = "GrebackGrotesquePERSONAL-Regular.otf"



font1 = fm.FontProperties(fname = path, size = 20)
font2 = fm.FontProperties(fname = path2, size = 12)
font3 = fm.FontProperties(fname = path3, size = 10)
font4 = fm.FontProperties(fname = path4, size = 10)
font5 = fm.FontProperties(fname = path2, size = 10)
```

```python
plt.figure(figsize = (10,5))


plt.barh(tahun, diabetes, color = '#219ebc', label = 'Diabetes Mellitus')
plt.barh(tahun, ginjal, color = '#4b5e6f', label = 'Chronic Kidney Disease', left = diabetes)
plt.barh(tahun, hepatitis, color = '#cbcbcb', label = 'Acute hepatitis', left = diabetes + ginjal)
plt.barh(tahun, HIV_AIDS, color = '#bb4747', label = 'HIV/AIDS', left = diabetes + ginjal + hepatitis)
plt.barh(tahun, malaria, color = '#7c8a62', label = 'Malaria', left = diabetes + ginjal + hepatitis + HIV_AIDS)


plt.title('Causes of Deaths in Indonesia\n(Age between 15 - 49)', fontproperties = font1)
plt.xlabel('Count', fontproperties = font2)
plt.ylabel('Year', fontproperties = font2)

plt.legend(prop = font2)

plt.yticks(np.arange(1990, 2022, 5), fontproperties = font4)
plt.xticks(fontproperties = font4)

plt.grid(linestyle = 'dashed', axis = 'x')

plt.show()
```
