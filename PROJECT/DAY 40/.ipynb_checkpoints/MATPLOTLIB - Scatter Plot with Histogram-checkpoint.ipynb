```python
import matplotlib.pyplot as plt
from matplotlib.gridspec import GridSpec
import pandas as pd
import numpy as np
```

```python
file = "energy.csv"

df = pd.read_csv(file)

df
```

```python
world = df.loc[df['Country'] == 'World']

world
```

```python
country = df.set_index('Country')

country
```

```python
world = country.drop('World')

world
```

```python
cons = world['CO2_emission']

cons
```

```python
gdp = world['GDP']

gdp
```

```python
plt.scatter(cons, gdp)


plt.show()
```

```python
fig = plt.figure()
gs = GridSpec(4, 4)

plt_scatter = fig.add_subplot(gs[1:4, 0:3])
plt_hist_y = fig.add_subplot(gs[0,0:3])
plt_hist_x = fig.add_subplot(gs[1:4, 3])

plt_scatter.scatter(cons, gdp)

plt_hist_x.hist(cons)
plt_hist_y.hist(gdp, orientation = 'horizontal')


plt.show()
```

```python
def scatter_hist(cons, gdp, plt, plt_histx, plt_histy):
    
    plt_histx.tick_params(axis="x", labelbottom=False)
    plt_histy.tick_params(axis="y", labelleft=False)

    plt.scatter(cons, gdp)

   
    plt_histx.hist(cons)
    plt_histy.hist(gdp, orientation='horizontal')


ax = fig.add_subplot(gs[1, 0])
ax_histx = fig.add_subplot(gs[0, 0], sharex=ax)
ax_histy = fig.add_subplot(gs[1, 1], sharey=ax)    

scatter_hist(cons, gdp, plt, ax_histx, ax_histy)
```

```python
file = "gapminder_data_graphs.csv"

df = pd.read_csv(file)

df
```

```python
year_2018 = df.loc[df['year'] == 2018]

year_2018
```

```python
life = year_2018['life_exp']

life
```

```python
hdi = year_2018['hdi_index']

hdi
```

```python
fig = plt.figure()
gs = GridSpec(4, 4)

plt_scatter = fig.add_subplot(gs[1:4, 0:3])
plt_bar_y = fig.add_subplot(gs[0,0:3])
plt_bar_x = fig.add_subplot(gs[1:4, 3])

plt_scatter.scatter(life, hdi)

plt_bar_x.hist(life, orientation = 'horizontal')
plt_bar_y.hist(hdi, orientation = 'vertical')


plt.show()
```
