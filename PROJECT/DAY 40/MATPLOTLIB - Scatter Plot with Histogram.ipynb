# MATPLOTLIB - Scatter Plot with Histogram

```python
import matplotlib.pyplot as plt
from matplotlib.gridspec import GridSpec
import pandas as pd
import numpy as np
```

```python
file = "gapminder_data_graphs.csv"

df = pd.read_csv(file)

df
```

```python
year_2018 = df.loc[df['year'] == 2018]

year_2018
```

```python
life18 = year_2018['life_exp']

life18 = life18.to_numpy()

life18
```

```python
year_2017 = df.loc[df['year'] == 2017]

year_2017
```

```python
life17 = year_2017['life_exp']

life17 = life17.to_numpy()

life17
```

```python
year_2016 = df.loc[df['year'] == 2016]

year_2016
```

```python
life16 = year_2016['life_exp']

life16 = life16.to_numpy()

life16
```

```python
hdi18 = year_2018['hdi_index']

hdi18
```

```python
gdp18 = year_2018['gdp']

gdp18
```

```python
hdi17 = year_2017['hdi_index']

hdi17
```

```python
gdp17 = year_2017['gdp']

gdp17
```

```python
hdi16 = year_2016['hdi_index']

hdi16
```

```python
gdp16 = year_2016['gdp']

gdp16
```

```python
import matplotlib.font_manager as fm
```

```python
path = "MADEAvenuePERSONALUSE-Regular.otf"
path2 = "Philosopher-Regular.ttf"
path3 = "Quicksand_Bold.otf"
path4 = "GrebackGrotesquePERSONAL-Regular.otf"



font1 = fm.FontProperties(fname = path, size = 23)
font2 = fm.FontProperties(fname = path2, size = 13)
font3 = fm.FontProperties(fname = path3, size = 10)
font4 = fm.FontProperties(fname = path4, size = 10)
font5 = fm.FontProperties(fname = path2, size = 10)
```

```python
fig = plt.figure(figsize = (8,6))
gs = GridSpec(4, 4)

ax1 = fig.add_subplot(gs[1:4, 0:3], facecolor = '#f7f9f2')
ax2 = fig.add_subplot(gs[0,0:3], facecolor = '#f7f9f2')
ax3 = fig.add_subplot(gs[1:4, 3], facecolor = '#f7f9f2')



ax1.scatter(life16, hdi16, color = '#941d26', label = 2016, s = 20)
ax1.scatter(life17, hdi17, color = '#406c35', label = 2017, s = 20)
ax1.scatter(life18, hdi18, color = '#dab678', label = 2018, s = 20)
ax1.legend(prop = font2)


ax1.grid(linestyle = 'dashed')
ax1.set_xlabel('Life Expectancy', fontproperties = font2)
ax1.set_ylabel('Human Development Index', fontproperties = font2)


fig.tight_layout(pad=1)

ax2.hist([life16, life17, life18], orientation = 'vertical', bins = 40, histtype = 'bar', color = ['#941d26','#406c35','#dab678'])
ax2.set_ylabel('Count', fontproperties = font2)
ax2.grid(linestyle = 'dashed', axis = 'y')

ax3.hist(hdi16, orientation = 'horizontal', bins = 40, stacked = True, histtype = 'bar', color = '#941d26', alpha = 0.5, edgecolor = 'black')
ax3.hist(hdi17, orientation = 'horizontal', bins = 40, stacked = True, histtype = 'bar', color = '#406c35', alpha = 0.5, edgecolor = 'black')
ax3.hist(hdi18, orientation = 'horizontal', bins = 40, stacked = True, histtype = 'bar', color = '#dab678', alpha = 0.5, edgecolor = 'black')
ax3.set_xlabel('Count', fontproperties = font2)
ax3.grid(linestyle = 'dashed', axis = 'x')

plt.show()
```
