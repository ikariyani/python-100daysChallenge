```python
tahun = np.arange(2010, 2020)

tahun
```

```python
import matplotlib.font_manager as fm
```

```python
path = "MADEAvenuePERSONALUSE-Regular.otf"
path2 = "Philosopher-Regular.ttf"
path3 = "Quicksand_Bold.otf"
path4 = "GrebackGrotesquePERSONAL-Regular.otf"



font1 = fm.FontProperties(fname = path, size = 22)
font2 = fm.FontProperties(fname = path2, size = 12.25)
font3 = fm.FontProperties(fname = path3, size = 10)
font4 = fm.FontProperties(fname = path4, size = 10)
font5 = fm.FontProperties(fname = path2, size = 16)
```

```python
plt.figure(figsize = (9,5))

keterangan = ['age <5', 'age 5-14', 'age 15-49', 'age 50-69', 'age >70']

plt.barh(tahun, usia1, label = 'age <5', color = '#c23119', zorder = 3)
plt.barh(tahun, usia2, left = usia1, label = 'age 5-14', color = '#d99a4e', zorder = 3)
plt.barh(tahun, usia3, left = usia1 + usia2, label = 'age 15-49', color = '#dadb9d', zorder = 3)
plt.barh(tahun, usia4, left = usia1 + usia2 + usia3, label = 'age 50-69', color = '#92bd54', zorder = 3)
plt.barh(tahun, usia5, left = usia1 + usia2 + usia3 + usia4, label = 'age >70', color = '#238a3c', zorder = 3)


plt.title('Cause of Death in Indonesia (Nutritional Deficiencies)', pad = 40, fontproperties = font1)
plt.xlabel('Total Death', fontproperties = font5)
plt.ylabel('Year', fontproperties = font5)

plt.xticks(np.arange(0, 23000, 2000), fontproperties = font4)
plt.yticks(np.arange(2010, 2020, 1), fontproperties = font4)

plt.grid(zorder = 0, axis = 'x', linestyle = 'dashed')
plt.legend(bbox_to_anchor=(-0.011,0.98),  loc='lower left', ncols = len(keterangan), prop = font2)

plt.show()
```
