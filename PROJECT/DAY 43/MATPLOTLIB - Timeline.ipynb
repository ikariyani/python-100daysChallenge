# MATPLOTLIB - Timeline

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
intro = 1
sql = 3
python = 8
numpy = 9
pandas = 8
matplotlib = 13
```

```python
belajar = np.array([1,3,8,9,8,13])

belajar
```

```python
materi = np.array(['Introduction', 'SQL', 'Basic Python', 'NumPy', 'Pandas', 'Matplotlib'])

materi
```

```python
import matplotlib.font_manager as fm
```

```python
path = "MADEAvenuePERSONALUSE-Regular.otf"
path2 = "Philosopher-Regular.ttf"
path3 = "Quicksand_Bold.otf"
path4 = "GrebackGrotesquePERSONAL-Regular.otf"



font1 = fm.FontProperties(fname = path, size = 23)
font2 = fm.FontProperties(fname = path2, size = 15)
font3 = fm.FontProperties(fname = path3, size = 10)
font4 = fm.FontProperties(fname = path4, size = 10)
font5 = fm.FontProperties(fname = path2, size = 12)
```

```python
# Cara 1

plt.figure(figsize = (10,3))
plt.rcParams['axes.facecolor'] = '#f7f9f2'

plt.barh(materi[0], belajar[0], color = '#daf1de',zorder = 2, label = 'Introduction')
plt.barh(materi[1], belajar[1], left = belajar[0], color = '#8eb69b', zorder = 2, label = 'SQL')
plt.barh(materi[2], belajar[2], left = belajar[0] + belajar[1], color = '#235347', zorder = 2, label = 'Basic Python')
plt.barh(materi[3], belajar[3], left = belajar[0] + belajar[1] + belajar [2], color = '#163832', zorder = 2, label = 'NumPy')
plt.barh(materi[4], belajar[4], left = belajar[0] + belajar[1] + belajar [2] + belajar[3], color = '#0b2b26', zorder = 2, label = 'Pandas')
plt.barh(materi[5], belajar[5], left = belajar[0] + belajar[1] + belajar [2] + belajar[3] + belajar [4], color = '#051f20', zorder = 2, label = 'Matplotlib')


plt.title('Timeline Belajar Python', fontproperties = font1)     
plt.xlabel('Hari ke-', fontproperties = font2)
plt.ylabel('Topik', fontproperties = font2)

plt.xticks(fontproperties = font5)
plt.yticks(fontproperties = font5)

plt.grid(axis = 'x', linestyle = 'dashed', zorder = 0)
plt.legend(loc = 'lower right',prop = font5)


plt.show()
```
