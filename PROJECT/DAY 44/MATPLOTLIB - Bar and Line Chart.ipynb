# MATPLOTLIB - Bar and Line Chart

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
hujan = np.array([300.7, 294.7, 210.8, 147.3, 132.1, 96.5, 63.5, 43.2, 66.0, 110.8, 142.2, 208.2])

hujan
```

```python
high = np.array([28.9, 28.9, 29.4, 30.0, 30.6, 30.0, 30.0, 30.6, 31.1, 31.1, 30.6, 29.4])

high
```

```python
low = np.array([23.3, 23.3, 23.3, 23.9, 23.9, 23.3, 22.8, 22.8, 22.8, 23.3, 23.3, 23.3])
```

```python
bulan = np.array(['Jan', 'Feb', 'Mar', 'Apr', 'Mei', 'Jun', 'Jul', 'Ags', 'Sept', 'Okt', 'Nov', 'Des'])
```

```python
import matplotlib.font_manager as fm
```

```python
path = "Galgony.ttf"
path2 = "MADEAvenuePERSONALUSE-Regular.otf"
path3 = "Megant√©-Personal-Use.ttf"




font1 = fm.FontProperties(fname = path, size = 18)
font2 = fm.FontProperties(fname = path2, size = 22)
font3 = fm.FontProperties(fname = path3, size = 10)
font4 = fm.FontProperties(fname = path, size = 10)
font5 = fm.FontProperties(fname = path2, size = 10)
```

```python
fig, ax1 = plt.subplots( figsize = (9,6))

ax1.bar(bulan, hujan, color = '#6a9798', edgecolor = '#206261', linewidth = 2, zorder = 3, label = 'Curah Hujan')
ax1.set_xlabel('Bulan', fontproperties = font1)
ax1.set_ylabel('Curah Hujan', fontproperties = font1)
plt.xticks(fontproperties = font4)
plt.yticks(np.arange(0, 400, 50))


plt.title('Diagram Curah Hujan dan Temperatur (Min - Maks)\nDKI Jakarta', fontproperties = font2)

ax2 = ax1.twinx()
ax2.plot(bulan, low, color = '#0b2b26', linestyle = 'dashed', marker= 'o', label = 'Temperatur Minimum')
ax2.set_ylabel('Suhu', fontproperties = font1)

ax2.plot(bulan, high, color = '#560c25', linestyle = 'dashed', marker= 'o', label = 'Temperatur Maksimum')


ax1.grid(linestyle = 'dashed', zorder = 0, axis = 'y')
plt.legend(prop = font5)

plt.show()
```
