```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('death-rates-from-air-pollution.csv')

df
```

```python
df_negara = df.loc[(df['Entity'] == 'Indonesia') |
                   (df['Entity'] == 'Malaysia') |
                   (df['Entity'] == 'Thailand') |
                   (df['Entity'] == 'Philippines') |
                    (df['Entity'] == 'Philippines')]

df_negara
```

```python
sns.jointplot(data=df_negara, x="Outdoor particulate matter (deaths per 100,000)", y="Indoor air pollution (deaths per 100,000)")

plt.show()
```

```python
sns.jointplot(data=df_negara, x="Outdoor particulate matter (deaths per 100,000)", y="Indoor air pollution (deaths per 100,000)", hue='Entity')

plt.show()
```

```python
sns.jointplot(data=df_negara, x="Outdoor particulate matter (deaths per 100,000)", y="Indoor air pollution (deaths per 100,000)", hue='Entity', kind='kde')

plt.show()
```

```python
sns.jointplot(data=df_negara, x="Outdoor particulate matter (deaths per 100,000)", y="Indoor air pollution (deaths per 100,000)", kind='reg')

plt.show()
```

```python
sns.jointplot(data=df_negara, x="Outdoor particulate matter (deaths per 100,000)", y="Indoor air pollution (deaths per 100,000)", kind='hist')

plt.show()
```

```python
sns.jointplot(data=df_negara, x="Outdoor particulate matter (deaths per 100,000)", y="Indoor air pollution (deaths per 100,000)", kind='hex')

plt.show()
```
