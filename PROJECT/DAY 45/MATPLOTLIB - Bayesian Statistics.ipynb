# MATPLOTLIB - Bayesian Statistics

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
import matplotlib.font_manager as fm
```

```python
path = "dream orphans.otf"
path2 = "FeFCrm2.ttf"
path3 = "steelfish rg.otf"
path4 = "Philosopher-Regular.ttf"


font1 = fm.FontProperties(fname = path, size = 13)
font2 = fm.FontProperties(fname = path2, size = 15)
font3 = fm.FontProperties(fname = path3, size = 10)
font4 = fm.FontProperties(fname = path4, size = 25)
font5 = fm.FontProperties(fname = path4, size = 15)
```

```python

np.random.seed(298746324)

plt.style.use('bmh')


def plot_beta_hist(ax, a, b, x):
    ax.hist(np.random.beta(a, b, size=5000),
            histtype="stepfilled", bins=30, alpha=0.8, density=True, edgecolor = 'black', linewidth = 1, label = x)


fig, ax = plt.subplots()
plot_beta_hist(ax, 7, 5, 'Data Keuangan')
plot_beta_hist(ax, 7, 16,  'Data Kependudukan')
plot_beta_hist(ax, 60, 10, 'Data Pejualan')
plot_beta_hist(ax, 6, 75, 'Data Lokasi')
ax.set_title("Probabilitas Pencurian Data\ndengan Metode Bayesian", fontproperties = font4, pad = 10)

plt.xlabel('Nilai Peluang', fontproperties = font5)
plt.ylabel('Densitas', fontproperties = font5)

plt.xticks(fontproperties = font1)
plt.yticks(fontproperties = font1)

plt.grid(axis = 'x')
plt.legend(prop = font5)

plt.show()
```
