# SEABORN - Joint Plot

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
file = "keparahan.csv"

df = pd.read_csv(file, sep=';')

df
```

```python
file2 = "D:\BELAJAR\GITHUB\python-100daysChallenge\PROJECT\DAY 46\kedalaman kemiskinan.csv"

df2 = pd.read_csv(file2, sep=';')

df2
```

```python
df = df[['nama_tahun', 'data_content', 'nama_item_vertical_variabel']]

df
```

```python
df2 = df2[['nama_tahun', 'data_content', 'nama_item_vertical_variabel']]

df2
```

```python
# tabel1 = df.set_index('nama_item_vertical_variabel')

# tabel1
```

```python
tabel1 = df.rename(columns={'nama_tahun' : 'tahun1', 'data_content' : 'keparahan', 'nama_item_vertical_variabel' : 'provinsi1'})

tabel1
```

```python
# tabel2 = df2.set_index('nama_item_vertical_variabel')

# tabel2
```

```python
tabel2 = df2.rename(columns={'nama_tahun' : 'tahun2', 'data_content' : 'kedalaman', 'nama_item_vertical_variabel' : 'provinsi2'})

tabel2
```

```python
join = tabel1.join(tabel2)

join
```

```python
sns.jointplot(data = join, x = 'keparahan', y = 'kedalaman', hue = 'provinsi1', kind = 'scatter')




plt.show()
```

```python
import matplotlib.font_manager as fm
```

```python
path = "MADEAvenuePERSONALUSE-Regular.otf"
path2 = "Philosopher-Regular.ttf"
path3 = "Quicksand_Bold.otf"
path4 = "GrebackGrotesquePERSONAL-Regular.otf"



font1 = fm.FontProperties(fname = path, size = 18)
font2 = fm.FontProperties(fname = path2, size = 13)
font3 = fm.FontProperties(fname = path3, size = 10)
font4 = fm.FontProperties(fname = path4, size = 10)
font5 = fm.FontProperties(fname = path2, size = 10)
```

```python


sns.jointplot(data = join, x = 'keparahan', y = 'kedalaman', hue = 'provinsi1', kind = 'scatter', marker = 'o', s=100, palette = ['#163832', '#d43d4f', '#7facbb'], zorder = 3)

x = join['keparahan']
y = join['kedalaman']


plt.title('Persebaran Indeks Kedalaman Kemiskinan\ndan Indeks Keparahan Kemiskinan', pad = 90, fontproperties = font1)
plt.xlabel('Indeks Keparahan Kemiskinan', fontproperties = font2)
plt.ylabel('Indeks Kedalaman Kemiskinan', fontproperties = font2)

plt.xticks(fontproperties = font3)
plt.yticks(fontproperties = font3)

plt.grid(linestyle = 'dotted', zorder = 0)

plt.legend(prop = font5)

plt.show()
```
