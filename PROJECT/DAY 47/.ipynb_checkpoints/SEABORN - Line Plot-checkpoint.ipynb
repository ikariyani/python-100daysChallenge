# SEABORN - ..............

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
file = "keparahan.csv"

df = pd.read_csv(file, sep=';')

df
```

```python
file2 = "D:\BELAJAR\GITHUB\python-100daysChallenge\PROJECT\DAY 46\kedalaman kemiskinan.csv"

df2 = pd.read_csv(file2, sep=';')

df2
```

```python
df = df[['nama_tahun', 'data_content', 'nama_item_vertical_variabel']]

df
```

```python
df2 = df2[['nama_tahun', 'data_content', 'nama_item_vertical_variabel']]

df2
```

```python
# tabel1 = df.set_index('nama_item_vertical_variabel')

# tabel1
```

```python
tabel1 = df.rename(columns={'nama_tahun' : 'tahun1', 'data_content' : 'keparahan', 'nama_item_vertical_variabel' : 'provinsi1'})

tabel1
```

```python
# tabel2 = df2.set_index('nama_item_vertical_variabel')

# tabel2
```

```python
tabel2 = df2.rename(columns={'nama_tahun' : 'tahun2', 'data_content' : 'kedalaman', 'nama_item_vertical_variabel' : 'provinsi2'})

tabel2
```

```python
join = tabel1.join(tabel2)

join
```

```python
sns.jointplot(data = join, x = 'keparahan', y = 'kedalaman', hue = 'provinsi1', kind = 'scatter')
plt.plot()




plt.show()
```

```python
sns.jointplot(data = join, x = 'keparahan', y = 'kedalaman', hue = 'provinsi1', kind = 'scatter', marker = 'o', s=100, palette = ['#163832', '#d43d4f', '#7facbb'], zorder = 3)

x = join['keparahan']
y = join['kedalaman']

plt.xlabel('Indeks Keparahan Kemiskinan')
plt.ylabel('Indeks Kedalaman Kemiskinan')

plt.grid(linestyle = 'dotted', zorder = 0)

plt.legend()

plt.show()
```

```python
fig, ax = plt.subplots(figsize = (10, 6))


sns.lineplot(ax = ax, data = join.loc[join['provinsi1'] == 'JAWA BARAT'], x = join['tahun1'], y = join['kedalaman'], palette = 'red', alpha = 0.8)
sns.lineplot(ax = ax, data = join.loc[join['provinsi1'] == 'JAWA TENGAH'], x = join['tahun1'], y = join['kedalaman'], palette = 'green', alpha = 0.5)
sns.lineplot(ax = ax, data = join.loc[join['provinsi1'] == 'JAWA TIMUR'], x = join['tahun1'], y = join['kedalaman'], palette = 'blue', alpha = 0.2)
```

```python
fig, ax = plt.subplots(figsize = (10, 6))

x = join['tahun1']
y = join['kedalaman']

sns.lineplot(ax = ax, data = join, x = x, y = y, palette = 'red')
ax.scatter(x.loc[join['provinsi1'] == 'JAWA BARAT'], y.loc[join['provinsi1'] == 'JAWA BARAT'], label = 'JAWA BARAT')
ax.scatter(x.loc[join['provinsi1'] == 'JAWA TENGAH'], y.loc[join['provinsi1'] == 'JAWA TENGAH'], label = 'JAWA TENGAH')
ax.scatter(x.loc[join['provinsi1'] == 'JAWA TIMUR'], y.loc[join['provinsi1'] == 'JAWA TIMUR'], label = 'JAWA TIMUR')

plt.legend()

plt.show()
```

```python
join['provinsi1'].unique()
```

```python
fig, ax = plt.subplots(figsize = (10, 6))

x = join['tahun1']
y = join['kedalaman']

sns.lineplot(ax = ax, data = join, x = x, y = y, palette = 'red')

prov = join['provinsi1'].unique()

for provinsi in prov :
    ax.scatter(x.loc[join['provinsi1'] == provinsi], y.loc[join['provinsi1'] == provinsi], label = provinsi)


plt.legend()

plt.show()
```
