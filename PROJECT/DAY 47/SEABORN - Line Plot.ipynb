# SEABORN - Line Plot

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
file = "keparahan.csv"

df = pd.read_csv(file, sep=';')

df
```

```python
tabel1 = df.rename(columns={'nama_tahun' : 'tahun1', 'data_content' : 'keparahan', 'nama_item_vertical_variabel' : 'provinsi1'})

tabel1
```

```python
# tabel2 = df2.set_index('nama_item_vertical_variabel')

# tabel2
```

```python
tabel2 = df2.rename(columns={'nama_tahun' : 'tahun2', 'data_content' : 'kedalaman', 'nama_item_vertical_variabel' : 'provinsi2'})

tabel2
```

```python
join = tabel1.join(tabel2)

join
```

```python
tahun = np.arange(2004, 2022, 1)

tahun
```

```python
join.loc[join['provinsi1'] == 'JAWA BARAT']
```

```python
fig, ax = plt.subplots(figsize = (10, 6))


sns.lineplot(ax = ax, data = join.loc[join['provinsi1'] == 'JAWA BARAT'], x = join['tahun1'], y = join['kedalaman'], palette = 'red', alpha = 0.8)
sns.lineplot(ax = ax, data = join.loc[join['provinsi1'] == 'JAWA TENGAH'], x = join['tahun1'], y = join['kedalaman'], palette = 'green', alpha = 0.5)
sns.lineplot(ax = ax, data = join.loc[join['provinsi1'] == 'JAWA TIMUR'], x = join['tahun1'], y = join['kedalaman'], palette = 'blue', alpha = 0.2)
```

```python
import matplotlib.font_manager as fm
```

```python
path = "MADEAvenuePERSONALUSE-Regular.otf"
path2 = "Philosopher-Regular.ttf"
path3 = "Quicksand_Bold.otf"
path4 = "GrebackGrotesquePERSONAL-Regular.otf"



font1 = fm.FontProperties(fname = path, size = 20)
font2 = fm.FontProperties(fname = path2, size = 13)
font3 = fm.FontProperties(fname = path3, size = 10)
font4 = fm.FontProperties(fname = path4, size = 10)
font5 = fm.FontProperties(fname = path2, size = 10)
```

```python
fig, ax = plt.subplots(figsize = (10, 6))

x = join['tahun1']
y = join['kedalaman']

sns.lineplot(ax = ax, data = join, x = x, y = y, color = '#235347', zorder = 3)
ax.plot(x.loc[join['provinsi1'] == 'JAWA BARAT'], y.loc[join['provinsi1'] == 'JAWA BARAT'], label = 'JAWA BARAT', color = '#1c4f6b',linestyle = 'dashed', marker = 'o', zorder = 3)
ax.plot(x.loc[join['provinsi1'] == 'JAWA TENGAH'], y.loc[join['provinsi1'] == 'JAWA TENGAH'], label = 'JAWA TENGAH', color = '#4f0205',linestyle = 'dashed', marker = 'o', zorder = 3)
ax.plot(x.loc[join['provinsi1'] == 'JAWA TIMUR'], y.loc[join['provinsi1'] == 'JAWA TIMUR'], label = 'JAWA TIMUR', color = '#ffbe00',linestyle = 'dashed', marker = 'o', zorder = 3)

plt.title('Persebaran Indeks Kedalaman Kemiskinan\ndan Indeks Keparahan Kemiskinan', pad = 20, fontproperties = font1)
plt.xlabel('Indeks Keparahan Kemiskinan', fontproperties = font2)
plt.ylabel('Indeks Kedalaman Kemiskinan', fontproperties = font2)

plt.xticks(np.arange(2004, 2021, 2),fontproperties = font3)
plt.yticks(fontproperties = font3)

plt.legend()
plt.grid(axis = 'y', linestyle = 'dashed', zorder = 0)

plt.show()
```
