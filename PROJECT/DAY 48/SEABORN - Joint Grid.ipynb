# SEABORN - Joint Grid

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
file2 = "D:\BELAJAR\GITHUB\python-100daysChallenge\PROJECT\DAY 48\energy.csv"
df2 = pd.read_csv(file2)

df2
```

```python
def kategori (x) :
    if (x == 'Indonesia' or x == 'Mexico' or x == 'Spain' or x == 'Turkey' or x == 'Netherlands') :
         return 'data'
    
    else :
        return 'non data'
    


df2['Kategori'] = df2['Country'].apply(kategori)
```

```python
data = df2.loc[df2['Kategori'] == 'data']

data
```

```python
data1 = data.loc[(data['Year'] == 2019) |
                 (data['Year'] == 2018) |
                 (data['Year'] == 2017) |
                 (data['Year'] == 2016) |
                 (data['Year'] == 2015) |
                 (data['Year'] == 2014) |
                 (data['Year'] == 2013)]
data1
```

```python
sns.scatterplot(data = data1, x = data1['Energy_production'],y = data1['Energy_consumption'], hue = 'Year', 
                style = 'Country', s= 70, palette = ['red', 'green', 'yellow', 'blue', 'orange'])
sns.rugplot(data=data1, x=data1['Energy_production'], y = data1['Energy_consumption'], height = 0.05)

plt.show()
```

```python
import matplotlib.font_manager as fm
```

```python
path = "MADEAvenuePERSONALUSE-Regular.otf"
path2 = "Philosopher-Regular.ttf"
path3 = "Quicksand_Bold.otf"
path4 = "GrebackGrotesquePERSONAL-Regular.otf"



font1 = fm.FontProperties(fname = path, size = 20)
font2 = fm.FontProperties(fname = path2, size = 15)
font3 = fm.FontProperties(fname = path3, size = 10)
font4 = fm.FontProperties(fname = path4, size = 10)
font5 = fm.FontProperties(fname = path2, size = 16)
```

```python

sns.set_style("whitegrid", {'grid.linestyle': '--'})
g = sns.JointGrid(data=data1, x=data1['Energy_production'], y=data1['Energy_consumption'], ratio=17, hue = 'Country')
g.plot_joint(sns.scatterplot,
             color="g", alpha=0.7, legend=font4, palette = ['#aa1803', '#bd613c', '#f1baa1', '#bcaf4d', '#6d8c00'], s = 120)
g.plot_marginals(sns.rugplot, height=1, alpha = 0.6, palette = ['#aa1803', '#bd613c', '#f1baa1', '#bcaf4d', '#6d8c00'])

g.set_axis_labels(xlabel = 'Energy Production', ylabel = 'Energy Consumption', fontproperties = font2)

plt.show()
```
