# SEABORN - Box Plot

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
file = "D:\BELAJAR\GITHUB\python-100daysChallenge\PROJECT\DAY 49\DATASET_cultural dimension of food consumption.xlsx"

df = pd.read_excel(file)

df
```

```python
df.head(2).transpose()
```

```python
df.columns
```

```python
df['(A1) CITY'].unique()
```

```python
df['income'] = df['(C12) HOUSEHOLD INCOME (MONTHLY IN MILLION IDR)']*1000000

df
```

```python
df['pengeluaran'] = df['(C14) HOUSEHOLD EXPENDITURE (MONTHLY IN MILLION IDR)']*1000000

df
```

```python
df['pengeluaran_makan'] = df['pengeluaran']*df['(C15) % MONTHLY EXPENDITURE FOR FOOD']/100

df
```

```python
penghasilan = df[['(A1) CITY', '(B3) GENDER', '(B4) AGE', '(B7) EDUCATION LEVEL', '(B9) OCCUPATION', 'income', 'pengeluaran', 'pengeluaran_makan' ]]

penghasilan
```

```python
box = penghasilan['income'], penghasilan['pengeluaran'], penghasilan['pengeluaran_makan']

box
```

```python
import matplotlib.font_manager as fm
```

```python
path = "MADEAvenuePERSONALUSE-Regular.otf"
path2 = "Philosopher-Regular.ttf"
path3 = "Quicksand_Bold.otf"
path4 = "GrebackGrotesquePERSONAL-Regular.otf"



font1 = fm.FontProperties(fname = path, size = 20)
font2 = fm.FontProperties(fname = path2, size = 13)
font3 = fm.FontProperties(fname = path3, size = 10)
font4 = fm.FontProperties(fname = path4, size = 10)
font5 = fm.FontProperties(fname = path2, size = 10)
```

```python
plt.boxplot([penghasilan['income'], penghasilan['pengeluaran'], penghasilan['pengeluaran_makan']], labels = ['Penghasilan', 'Pengeluaran', 'Pengeluaran Makan'])


plt.show()
```

```python
sns.boxplot(data = box, palette = 'flare')

label = ['Penghasilan', 'Pengeluaran', 'Pengeluaran Makan']


plt.title('Statistik Tingkat Konsumsi dan Penghasilan\nMasyarakat Perkotaan di Indonesia', fontproperties = font1, pad = 20)

plt.ylabel('Rupiah',  fontproperties = font2)
plt.xticks([0,1,2], labels = label, fontproperties = font2)
plt.yticks(fontproperties = font5)


plt.show()
```
