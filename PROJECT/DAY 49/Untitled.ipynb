```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
file = "energy.csv"

df = pd.read_csv(file)

df
```

```python
def kategori (x) :
    if (x == 'Indonesia' or x == 'Mexico' or x == 'Spain' or x == 'Turkey' or x == 'Netherlands') :
         return 'data'
    
    else :
        return 'non data'
    


df['Kategori'] = df['Country'].apply(kategori)
```

```python
data = df.loc[df['Kategori'] == 'data']

data
```

```python
data1 = data.loc[(data['Year'] == 2019)]

data1
```

```python
data2 = data.loc[
                 (data['Year'] == 2018)]

data2
```

```python
data1 = data1.loc[(data1['Country'] == 'Indonesia') ]

data1
```

```python
sns.boxplot(data = data1, x =  data1['GDP'])
sns.boxplot(data = data2, x =  data2['Energy_consumption'])
```
