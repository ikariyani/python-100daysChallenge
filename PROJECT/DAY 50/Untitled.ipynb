# SEABORN - Joint Plot

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
file = "D:\BELAJAR\GITHUB\python-100daysChallenge\PROJECT\DAY 50\WorldHappiness_Corruption_2015_2020.csv"

df = pd.read_csv(file)

df
```

```python
data = df.loc[(df['Year'] == 2020)  ]

data
```

```python
sns.jointplot(data = data, y = data['freedom'], x = data['health'], hue = 'Year', kind = 'kde', palette = ['#aa6f73', '#254a7f', '#f3b800', '#c23119'])
```

```python
import matplotlib.font_manager as fm
```

```python
path = "MADEAvenuePERSONALUSE-Regular.otf"
path2 = "Philosopher-Regular.ttf"
path3 = "Quicksand_Bold.otf"
path4 = "GrebackGrotesquePERSONAL-Regular.otf"



font1 = fm.FontProperties(fname = path, size = 17)
font2 = fm.FontProperties(fname = path2, size = 15)
font3 = fm.FontProperties(fname = path3, size = 10)
font4 = fm.FontProperties(fname = path4, size = 12)
font5 = fm.FontProperties(fname = path2, size = 16)
```

```python

sns.jointplot(data = data, x = data['government_trust'], y = data['health'], hue = 'Year', kind = 'kde', palette = ['#235347'], legend = font2)

plt.title('Kontribusi Kesehatan dan Kepercayaan terhadap Pemerintah\ndalam Indeks Kebahagiaan', pad = 80, fontproperties = font1)
plt.xlabel('Kepercayaan terhadap Pemerintah', fontproperties = font4)
plt.ylabel('Kesehatan', fontproperties = font4)

plt.xticks(fontproperties = font3)
plt.yticks(fontproperties = font3)

plt.grid(linestyle = 'dotted', zorder = 0)

plt.show()
```
