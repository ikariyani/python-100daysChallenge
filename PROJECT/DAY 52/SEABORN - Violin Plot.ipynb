# SEABORN - Violin Plot

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
file = "D:\BELAJAR\GITHUB\python-100daysChallenge\PROJECT\DAY 51\WorldHappiness_Corruption_2015_2020.csv"

df = pd.read_csv(file)

df
```

```python
data = df['gdp_per_capita'], df['family'], df['health'], df['freedom'], df['generosity'], df['government_trust'], df['dystopia_residual'], df['social_support']

data
```

```python
import matplotlib.font_manager as fm
```

```python
path = "Amelaryas.otf"
path2 = "AUBREY1__.TTF"
path3 = "Caver Pitalsal.otf"
path4 = "Galgony.ttf"


font1 = fm.FontProperties(fname = path, size = 10)
font2 = fm.FontProperties(fname = path4, size = 13)
font3 = fm.FontProperties(fname = path3, size = 30)
font4 = fm.FontProperties(fname = path4, size = 25)
font5 = fm.FontProperties(fname = path4, size = 13)
font6 = fm.FontProperties(fname = path2, size = 10)
```

```python
fig, ax = plt.subplots(figsize = (15, 6))
label = ['GDP per Capita', 'Family', 'Health', 'Freedom', 'Generosity', 'Government Trust', 'Dystopia Residual', 'Social Support']

sns.violinplot(ax = ax, data = data, palette = 'Spectral')

plt.title('Kontribusi Indikator Penentu Indeks Kebahagiaan', fontproperties = font4)
plt.xticks([0,1,2,3,4,5,6,7], labels = label, fontproperties = font2)
plt.yticks(fontproperties = font2)

plt.grid(axis = 'y', zorder = 0, linestyle = 'dashed', alpha = 0.4)

plt.show()
```
