# SEABORN - Dis Plot

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
file2 = "D:\BELAJAR\GITHUB\python-100daysChallenge\PROJECT\DAY 53\death-rates-from-air-pollution.csv"

df2 = pd.read_csv(file2)

df2
```

```python
df2_negara = df2.loc[(df2['Entity'] == 'Indonesia') |
                   (df2['Entity'] == 'Mexico') |
                   (df2['Entity'] == 'Netherlands') |
                   (df2['Entity'] == 'Spain') |
                    (df2['Entity'] == 'Turkey')]

df2_negara
```

```python
coba = df2.loc[(df2['Year'] == 2017) ]
```

```python
import matplotlib.font_manager as fm
```

```python
path = "MADEAvenuePERSONALUSE-Regular.otf"
path2 = "Philosopher-Regular.ttf"
path3 = "Quicksand_Bold.otf"
path4 = "GrebackGrotesquePERSONAL-Regular.otf"



font1 = fm.FontProperties(fname = path, size = 18)
font2 = fm.FontProperties(fname = path2, size = 13)
font3 = fm.FontProperties(fname = path3, size = 10)
font4 = fm.FontProperties(fname = path4, size = 12)
font5 = fm.FontProperties(fname = path2, size = 16)
```

```python
sns.displot(data = coba, x="Outdoor particulate matter (deaths per 100,000)", y="Indoor air pollution (deaths per 100,000)", 
            palette = 'rocket')

plt.title('Kasus Kematian Akibat Polusi Udara di Beberapa Negara\npada Tahun 2017', fontproperties = font1, pad = 20)
plt.xticks(fontproperties = font3)
plt.yticks(fontproperties = font3)

plt.xlabel('Outdoor particulate matter (deaths per 100,000)', fontproperties = font2)
plt.ylabel('Indoor air pollution (deaths per 100,000)', fontproperties = font2)

plt.grid(linestyle = 'dashed')

plt.show()
```
