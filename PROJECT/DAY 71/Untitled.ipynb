```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
file = "D:\BELAJAR\GITHUB\python-100daysChallenge\PROJECT\DAY 71\World Crime Index .csv"

df = pd.read_csv(file)

df
```

```python
def kategori (x) :
    if (x in ('Jakarta, Indonesia', 'Kuala Lumpur, Malaysia', 'Singapore, Singapore', 'Manila, Philippines', 'Bangkok, Thailand', 'Phnom Penh, Cambodia', 'Hanoi, Vietnam')) :
         return 'Asia Tenggara'
    
    else :
        return 'Non Asia Tenggara'
    


df['Kategori'] = df['City'].apply(kategori)
```

```python
data = df.loc[df['Kategori'] == 'Asia Tenggara']
```

```python
data.to_csv('D:\BELAJAR\GITHUB\python-100daysChallenge\PROJECT\DAY 71\crimedata.csv')
```

```python
def negara (x) : 
    if (x == 'Manila, Philippines') :
        return 'Philippines'
    elif (x == 'Jakarta, Indonesia') :
        return 'Indonesia'
    elif (x == 'Kuala Lumpur, Malaysia') :
        return 'Malaysia'
    elif (x == 'Bangkok, Thailand') :
        return 'Thailand'
    elif (x == 'Phnom Penh, Cambodia') :
        return 'Cambodia'
    elif (x == 'Singapore, Singapore') :
        return 'Singapore'
    elif (x == 'Hanoi, Vietnam') :
        return 'Vietnam'
    else :
        return 'Asia'


data['Country'] = data['City'].apply(negara)
```
