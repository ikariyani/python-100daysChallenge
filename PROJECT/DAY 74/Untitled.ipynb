# Membuat Fungsi Restrukturisasi

```python
df1['year'] = 0

df1
```

```python
year_2020 = df1[['provinsi', 'year', 'ks_2020', 'hp_2020', 'ld_2020']]

year_2020
```

```python
year_2020['year'] = 2020

year_2020
```

```python
year_2020 = year_2020.rename(columns = {'ks_2020' : 'kebebasan_sipil', 'hp_2020' : 'hak_politik', 'ld_2020' : 'lembaga_demokrasi'})

year_2020
```

```python
def restrukturisasi (df, ks, hp, ld, tahun):
    df['year'] = 0
    year = df[['provinsi', 'year', ks, hp, ld]]
    year['year'] = tahun
    year = year.rename(columns = {ks : 'kebebasan_sipil', hp : 'hak_politik', ld : 'lembaga_demokrasi'})
    return year
```

```python
tabel = None
tahun = np.arange(2011, 2021, 1)

for x in tahun :
    print(x)
```

```python
year_2020 = restrukturisasi(df1, 'ks_2020', 'hp_2020', 'ld_2020', 2020)
year_2019 = restrukturisasi(df1, 'ks_2019', 'hp_2019', 'ld_2019', 2019)
year_2018 = restrukturisasi(df1, 'ks_2018', 'hp_2018', 'ld_2018', 2018)
year_2017 = restrukturisasi(df2, 'ks_2017', 'hp_2017', 'ld_2017', 2017)
year_2016 = restrukturisasi(df2, 'ks_2016', 'hp_2016', 'ld_2016', 2016)
year_2015 = restrukturisasi(df2, 'ks_2015', 'hp_2015', 'ld_2015', 2015)
year_2014 = restrukturisasi(df3, 'ks_2014', 'hp_2014', 'ld_2014', 2014)
year_2013 = restrukturisasi(df3, 'ks_2013', 'hp_2013', 'ld_2013', 2013)
year_2012 = restrukturisasi(df3, 'ks_2012', 'hp_2012', 'ld_2012', 2012)
year_2011 = restrukturisasi(df4, 'ks_2011', 'hp_2011', 'ld_2011', 2011)
```

```python
tabel = pd.concat([year_2020, year_2019, year_2018, year_2017, year_2016, year_2015, year_2014, year_2013, year_2012, year_2011])

tabel
```

```python
import seaborn as sns
```

```python
import matplotlib.font_manager as fm
```

```python
path = "MADEAvenuePERSONALUSE-Regular.otf"
path2 = "Philosopher-Regular.ttf"
path3 = "Quicksand_Bold.otf"
path4 = "GrebackGrotesquePERSONAL-Regular.otf"



font1 = fm.FontProperties(fname = path, size = 20)
font2 = fm.FontProperties(fname = path2, size = 13)
font3 = fm.FontProperties(fname = path3, size = 10)
font4 = fm.FontProperties(fname = path4, size = 10)
font5 = fm.FontProperties(fname = path2, size = 10)
```

```python
plt.figure(figsize = [10,5])

sns.boxplot(data = tabel, x = 'provinsi', y = 'kebebasan_sipil', palette = 'rocket')

plt.title('Boxplot Aspek Kebebasan Sipil Seluruh Provinsi di Indonesia\nTahun 2011 - 2020', fontproperties = font1)
plt.xlabel(None)
plt.ylabel('Kebebasan Sipil', fontproperties = font2)

plt.xticks(rotation = 90, fontproperties = font3)
plt.yticks(fontproperties = font3)
plt.grid(linestyle = 'dotted')

plt.show()
```

```python
plt.figure(figsize = [10,5])

sns.boxplot(data = tabel, x = 'provinsi', y = 'hak_politik', palette = 'crest')

plt.title('Boxplot Aspek Hak - Hak Politik Seluruh Provinsi di Indonesia\nTahun 2011 - 2020', fontproperties = font1)
plt.xlabel(None)
plt.ylabel('Hak - Hak Politik', fontproperties = font2)

plt.xticks(rotation = 90, fontproperties = font3)
plt.yticks(fontproperties = font3)
plt.grid(linestyle = 'dotted')

plt.show()
```
