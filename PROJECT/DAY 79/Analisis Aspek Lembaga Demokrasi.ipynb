```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
file = "indeks demokrasi.csv"

df = pd.read_csv(file, sep = ';')

df
```

```python
df = df.drop(columns = 'Unnamed: 0')

df
```

```python
IDI = df.loc[df['provinsi'] == 'INDONESIA']

IDI
```

```python
IDI = IDI.sort_values(by = ['year'], ascending = True)

IDI
```

```python
IDI['IDI'] = IDI[['kebebasan_sipil', 'hak_politik', 'lembaga_demokrasi']].mean(axis = 1)

IDI
```

```python
import matplotlib.font_manager as fm
```

```python
path = "MADEAvenuePERSONALUSE-Regular.otf"
path2 = "Philosopher-Regular.ttf"
path3 = "Quicksand_Bold.otf"
path4 = "GrebackGrotesquePERSONAL-Regular.otf"



font1 = fm.FontProperties(fname = path, size = 23)
font2 = fm.FontProperties(fname = path2, size = 13)
font3 = fm.FontProperties(fname = path3, size = 10)
font4 = fm.FontProperties(fname = path4, size = 10)
font5 = fm.FontProperties(fname = path2, size = 9)
```

```python
plt.figure(figsize = [9,5])

plt.bar(IDI['year'] , IDI['IDI'], color = '#8b9d77')
plt.plot(IDI['year'], IDI['kebebasan_sipil'], label = 'Kebebasan Sipil', marker = 'X', color = '#aa1803')
plt.plot(IDI['year'], IDI['hak_politik'], label = 'Hak - Hak Politik',  marker = 'X', color = '#344c11')
plt.plot(IDI['year'], IDI['lembaga_demokrasi'], label = 'Lembaga Demokrasi',  marker = 'X', color = '#ffbe00')

plt.title('Indeks Demokrasi Indonesia Tahun 2011 - 2020\ndan Aspek Penyusunnya', fontproperties = font1)

plt.xlabel('Tahun', fontproperties = font2)
plt.ylabel('Indeks Demokrasi Indonesia', fontproperties = font2)

plt.xticks(np.arange(2011, 2021, 1), fontproperties = font3)
plt.yticks(np.arange(0, 110, 10), fontproperties = font3)

plt.legend(loc = 'upper left', prop = font5)
plt.grid(axis = 'y', linestyle = 'dotted')

plt.show()
```
