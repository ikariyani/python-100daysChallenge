```python
import matplotlib.font_manager as fm
```

```python
path = "MADEAvenuePERSONALUSE-Regular.otf"
path2 = "Philosopher-Regular.ttf"
path3 = "Quicksand_Bold.otf"
path4 = "GrebackGrotesquePERSONAL-Regular.otf"



font1 = fm.FontProperties(fname = path, size = 25)
font2 = fm.FontProperties(fname = path2, size = 13)
font3 = fm.FontProperties(fname = path3, size = 9)
font4 = fm.FontProperties(fname = path4, size = 10)
font5 = fm.FontProperties(fname = path2, size = 15.98)
```

```python
sns.jointplot(data = df, y = df['kardio'], x = df['total consumption'], height = 7, kind = 'reg', color = '#7f4042')

plt.title('Korelasi Antara Kematian Akibat Kardiovaskuler\ndengan Total Konsumsi Daging di Indonesia\nTahun 1990 - 2017', pad = 90, fontproperties = font1)
plt.xlabel('Per Capita Meat Consumption\n(kg/year)', fontproperties = font2)
plt.ylabel('Cardiovascular Diseases', fontproperties = font2)

plt.xticks(fontproperties = font3)
plt.yticks(fontproperties = font3)

plt.grid(linestyle = 'dotted')

plt.show()
```

```python
sns.set({'figure.figsize':(16,5)})
sns.set_theme(style='white')

sns.barplot(data = df, x = df['Year'], y = df['total consumption'], palette = 'YlOrBr')

plt.title('Korelasi Antara Kematian Akibat Kardiovaskuler dengan Total Konsumsi Daging di Indonesia\nTahun 1990 - 2017', pad = 20, fontproperties = font1)
plt.xlabel('Per Capita Meat Consumption (kg/year)', fontproperties = font2)
plt.ylabel('Cardiovascular Diseases', fontproperties = font2)

plt.xticks(fontproperties = font3)
plt.yticks(fontproperties = font3)

plt.grid(linestyle = 'dotted', axis = 'y')

plt.show()
```

```python
plt.figure(figsize = (14,6))

keterangan = ['Domba', 'Unggas', 'Babi', 'Sapi', 'Lainnya']

plt.barh(df['Year'], b, label = 'Daging Domba', color = '#238a3c')
plt.barh(df['Year'], c, left = b, label = 'Daging Unggas', color = '#92bd54')
plt.barh(df['Year'], d, left = b + c, label = 'Daging Babi', color = '#dadb9d')
plt.barh(df['Year'], e, left = b + c + d, label = 'Daging Sapi', color = '#c23119')
plt.barh(df['Year'], f, left = b + c + d + e, label = 'Lainnya', color = '#d99a4e')

plt.title('Tingkat Konsumsi Beberapa Jenis Daging di Indonesia Tahun 1990 - 2017', pad = 50, fontproperties = font1)
plt.xlabel('Per Capita Meat Consumption (%)', fontproperties = font2)
plt.ylabel('Year', fontproperties = font2)

plt.xticks(fontproperties = font3)
plt.yticks(np.arange(1990, 2018,2), fontproperties = font3)

plt.legend(bbox_to_anchor=(-0.011,0.99),  loc='lower left', ncols = len(keterangan), prop = font5)
plt.grid(linestyle = 'dotted', axis = 'y')

plt.show()
```
