```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
file = "meat.csv"

df = pd.read_csv(file, sep = ';')

df
```

```python
unggas_con = df['Poultry (kilograms per year)']

unggas_con
```

```python
unggas_prod = df['poultry (Ton)']

unggas_prod
```

```python
unggas_con.corr(unggas_prod)
```

```python
sapi_con = df['Beef and bufallo (kilograms per year)']

sapi_con
```

```python
sapi_prod = df['beef and buffalo (Ton)']

sapi_prod
```

```python
df['total production'] = df[['beef and buffalo (Ton)', 'poultry (Ton)', 'pigmeat (Ton)']].sum(axis = 1)

df
```

```python
import matplotlib.font_manager as fm
```

```python
path = "MADEAvenuePERSONALUSE-Regular.otf"
path2 = "Philosopher-Regular.ttf"
path3 = "Quicksand_Bold.otf"
path4 = "GrebackGrotesquePERSONAL-Regular.otf"



font1 = fm.FontProperties(fname = path, size = 25)
font2 = fm.FontProperties(fname = path2, size = 13)
font3 = fm.FontProperties(fname = path3, size = 9)
font4 = fm.FontProperties(fname = path, size = 20)
font5 = fm.FontProperties(fname = path2, size = 15.98)
```

```python
sapi_con.corr(sapi_prod)
```

```python
plt.figure(figsize = (10,4))

plt.bar(df['Year'] , df['beef and buffalo (Ton)'], color = '#344c11')


plt.title('Tingkat Produksi Daging Tahun 1990 - 2017', pad = 20, fontproperties = font1)
plt.xlabel('Tahun', fontproperties = font2)
plt.ylabel('Produksi Daging', fontproperties = font2)

plt.xticks(fontproperties = font3)
plt.yticks(fontproperties = font3)

plt.grid(linestyle = 'dotted')

plt.show()

```

```python
sapi = df['beef and buffalo (Ton)'] / df['total production'] * 100

sapi
```

```python
unggas = df['poultry (Ton)'] / df['total production'] * 100

unggas
```

```python
pigmeat = df['pigmeat (Ton)'] / df['total production'] * 100

pigmeat
```

```python
plt.figure(figsize = (14,6))

keterangan = ['Unggas', 'Babi', 'Sapi']

plt.barh(df['Year'], sapi, label = 'Daging Sapi/Kerbau', color = '#406c35')
plt.barh(df['Year'], unggas, left = sapi, label = 'Daging Unggas', color = '#ceba9f')
plt.barh(df['Year'], pigmeat, left = sapi + unggas, label = 'Daging Babi', color = '#4f0205')


plt.title('Tingkat Produksi Beberapa Jenis Daging di Indonesia Tahun 1990 - 2017', pad = 50, fontproperties = font1)
plt.xlabel('Meat Production (Ton)', fontproperties = font2)
plt.ylabel('Tahun', fontproperties = font2)

plt.xticks(fontproperties = font3)
plt.yticks(np.arange(1990, 2018,2), fontproperties = font3)

plt.legend(bbox_to_anchor=(0.5,1.05),  loc='center', ncols = len(keterangan), prop = font5)
plt.grid(linestyle = 'dotted', axis = 'y')

plt.show()
```

```python
sns.jointplot(data = df, x = df['Poultry (kilograms per year)'], y = df['poultry (Ton)'], kind = 'reg', color = '#0b3453')

plt.title('Korelasi Antara Tingkat Konsumsi\ndengan Tingkat Produksi Daging Unggas di Indonesia\nTahun 1990 - 2017', pad = 80, fontproperties = font4)
plt.xlabel('Per Capita Poultry Consumption\n(kg/year)', fontproperties = font2)
plt.ylabel('Poultry Production (Ton)', fontproperties = font2)

plt.xticks(fontproperties = font3)
plt.yticks(fontproperties = font3)

plt.grid(linestyle = 'dotted')

plt.show()

```

```python
sns.jointplot(data = df, x = df['Beef and bufallo (kilograms per year)'], y = df['beef and buffalo (Ton)'], kind = 'reg', color = '#3d2418')

plt.title('Korelasi Antara Tingkat Konsumsi\ndengan Tingkat Produksi Daging Sapi/Kerbau di Indonesia\nTahun 1990 - 2017', pad = 80, fontproperties = font4)
plt.xlabel('Per Capita Beef and Bufallo Consumption\n(kg/year)', fontproperties = font2)
plt.ylabel('Beef and Bufallo Production (Ton)', fontproperties = font2)

plt.xticks(fontproperties = font3)
plt.yticks(fontproperties = font3)

plt.grid(linestyle = 'dotted')

plt.show()

```
