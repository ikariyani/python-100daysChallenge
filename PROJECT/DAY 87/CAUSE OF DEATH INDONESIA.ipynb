```python
import matplotlib.font_manager as fm
```

```python
path = "MADEAvenuePERSONALUSE-Regular.otf"
path2 = "Philosopher-Regular.ttf"
path3 = "Quicksand_Bold.otf"
path4 = "GrebackGrotesquePERSONAL-Regular.otf"



font1 = fm.FontProperties(fname = path, size = 25)
font2 = fm.FontProperties(fname = path2, size = 13)
font3 = fm.FontProperties(fname = path3, size = 9)
font4 = fm.FontProperties(fname = path, size = 20)
font5 = fm.FontProperties(fname = path2, size = 12)
```

```python
plt.figure(figsize = (14,6))

keterangan = ['Cardiovascular Disease', 'Neoplasm', 'Digestive Disease', 'Tuberculosis', 'Diarrheal Diseases', 'Others']

plt.barh(indo['Year'], indo14, label = 'Cardiovascular Disease', color = '#500302')
plt.barh(indo['Year'], indo22, left = indo14, label = 'Neoplasm', color = '#a22e2e')
plt.barh(indo['Year'], indo31, left = indo14 + indo22, label = 'Digestive Disease', color = '#e0d59e')
plt.barh(indo['Year'], indo13, left = indo14 + indo22 + indo31, label = 'Tuberculosis', color = '#748a5e')
plt.barh(indo['Year'], indo20, left = indo14 + indo22 + indo31 + indo13, label = 'Diarrheal Diseases', color = '#4b7554')
plt.barh(indo['Year'], indolainnya, left = indo14 + indo22 + indo31 + indo13 + indo20, label = 'Others', color = '#365301')


plt.title('Kasus Kematian Akibat Beberapa Penyakit di Indonesia Tahun 1990 - 2019', pad = 40, fontproperties = font1)
plt.xlabel('Death (%)', fontproperties = font2)
plt.ylabel('Tahun', fontproperties = font2)

plt.xticks(fontproperties = font3)
plt.yticks(np.arange(1990, 2020,1), fontproperties = font3)

plt.legend(bbox_to_anchor=(0.5,1.05),  loc='center', ncols = len(keterangan), prop = font5)
plt.grid(linestyle = 'dotted', axis = 'x')

plt.show()
```
