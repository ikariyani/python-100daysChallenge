```python
import matplotlib.font_manager as fm
```

```python
path = "MADEAvenuePERSONALUSE-Regular.otf"
path2 = "Philosopher-Regular.ttf"
path3 = "Quicksand_Bold.otf"
path4 = "GrebackGrotesquePERSONAL-Regular.otf"



font1 = fm.FontProperties(fname = path, size = 25)
font2 = fm.FontProperties(fname = path2, size = 13)
font3 = fm.FontProperties(fname = path3, size = 9)
font4 = fm.FontProperties(fname = path, size = 22)
font5 = fm.FontProperties(fname = path2, size = 12)
font6 = fm.FontProperties(fname = path2, size = 18)
```

```python
plt.figure(figsize = (14,4))

plt.bar(india['Year'], india['Total'], color = '#445f3b', label = 'India')
plt.bar(indo['Year'], indo['Total'], color = '#8ea076', label = 'Indonesia')

plt.title('Tren Mental Health Disorder di India dan Indonesia Tahun 1990 - 2019', fontproperties = font1)
plt.xlabel('Year', fontproperties = font2)
plt.ylabel('Mental Health Disorder (%)', fontproperties = font2)


plt.xticks(np.arange(1990, 2020, 2), fontproperties = font3)
plt.yticks(np.arange(0, 13, 1), fontproperties = font3)

plt.legend(prop = font5)

plt.show()
```

```python
plt.figure(figsize = (9,4))

plt.stem(indo['Year'], indo['Total'], label = 'Indonesia', basefmt = '#084c61', linefmt = '#084c61')

plt.title('Tren Mental Health Disorder di Indonesia Tahun 1990 - 2019', fontproperties = font6)
plt.xlabel('Year', fontproperties = font2)
plt.ylabel('Mental Health Disorder (%)', fontproperties = font2)


plt.xticks(np.arange(1990, 2022, 2), fontproperties = font3)
plt.yticks(np.arange(0, 11, 1), fontproperties = font3)

plt.legend(prop = font5)

plt.show()
```
