```python
import matplotlib.font_manager as fm
```

```python
path = "MADEAvenuePERSONALUSE-Regular.otf"
path2 = "Philosopher-Regular.ttf"
path3 = "Quicksand_Bold.otf"
path4 = "GrebackGrotesquePERSONAL-Regular.otf"



font1 = fm.FontProperties(fname = path, size = 25)
font2 = fm.FontProperties(fname = path2, size = 13)
font3 = fm.FontProperties(fname = path3, size = 9)
font4 = fm.FontProperties(fname = path, size = 22)
font5 = fm.FontProperties(fname = path2, size = 12)
font6 = fm.FontProperties(fname = path2, size = 18)
```

```python
plt.figure(figsize = (14,4))

plt.bar(india['Year'], india['population'], color = '#1c4f6b', label = 'India')
plt.bar(indo['Year'], indo['population'], color = '#7895a2', label = 'Indonesia')

plt.title('Perkembangan Populasi di India dan Indonesia Tahun 1990 - 2019', fontproperties = font1)
plt.xlabel('Year', fontproperties = font2)
plt.ylabel('Population (jt)', fontproperties = font2)


plt.xticks(np.arange(1990, 2020, 2), fontproperties = font3)
plt.yticks(fontproperties = font3)

plt.legend(prop = font5)

plt.show()
```

```python
sns.jointplot(data = indo, y = indo['Total'], x = indo['population'], kind = 'reg', color = '#560c25')

plt.title('Korelasi Antara Mental Health Disorder \ndengan Perkembangan Populasi di Indonesia\n Tahun 1990 - 2019', pad = 80, fontproperties = font4)
plt.xlabel('Population', fontproperties = font2)
plt.ylabel('Mental Health Disorder (%)', fontproperties = font2)

plt.xticks(fontproperties = font3)
plt.yticks(fontproperties = font3)

plt.grid(linestyle = 'dotted')

plt.show()
```

```python
sns.jointplot(data = indo, y = total_india, x = populasi_india, kind = 'reg', color = '#0b3453')

plt.title('Korelasi Antara Mental Health Disorder \ndengan Perkembangan Populasi di Indonesia\n Tahun 1990 - 2019', pad = 80, fontproperties = font4)
plt.xlabel('Population', fontproperties = font2)
plt.ylabel('Mental Health Disorder (%)', fontproperties = font2)

plt.xticks(fontproperties = font3)
plt.yticks(fontproperties = font3)

plt.grid(linestyle = 'dotted')

plt.show()
```
