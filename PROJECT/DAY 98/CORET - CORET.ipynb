```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
file = "prevalence-by-mental-and-substance-use-disorder.csv"

df = pd.read_csv(file)

df
```

```python
df = df.loc[(df['Entity'] == 'Indonesia') |
       (df['Entity'] == 'India')]

df
```

```python
df = df.drop(index= [2640, 2670])

df
```

```python
df['Total'] = df.iloc[:, [3, 4, 5, 6, 7, 8, 9]].sum(axis = 1)

df
```

```python
df['ID'] = df['Code'].str.cat(df['Year'].astype('str'))

df
```

```python
file2 = "populasi.csv"

df2 = pd.read_csv(file2, sep = ';')

df2
```

```python
df2 = df2.drop(index = [0,30])

df2
```

```python
df2.loc[df2['Country/Territory'] == 'India','Code'] = 'IND'
df2.loc[df2['Country/Territory'] == 'Indonesia','Code'] = 'IDN'
```

```python
df2['ID'] = df2['Code'].str.cat(df2['Year'].astype('str'))

df2
```

```python
df
```

```python
populasi = df2['population']
```

```python
merge = pd.merge(df,populasi, on = populasi, how = 'left')
pd.set_option('display.max_rows', None)

merge
```

```python
df = merge.drop(columns = ['key_0'])

df
```

```python
file3 = "gdp.csv"

df3 = pd.read_csv(file3, sep = ';')

df3
```

```python
df3.loc[df3['Country Name'] == 'India','ID'] = 'IND'
df3.loc[df3['Country Name'] == 'Indonesia','ID'] = 'IDN'

df3
```

```python
file4 = "unemployment.csv"

df4 = pd.read_csv(file4, sep = ';')

df4
```

```python
df4.loc[df4['Country'] == 'India', 'Code'] = 'IND'
df4.loc[df4['Country'] == 'Indonesia', 'Code'] = 'IDN'

df4
```

```python
df4['ID'] = df4['Code'].str.cat(df4['Year'].astype('str'))

df4.set_index('ID')
```

```python
df = df.set_index('ID').join(df4.set_index('ID'), how='left', lsuffix='_left', rsuffix='_right')

df
```

```python
df = df2.set_index('ID').join(df, how='left', lsuffix='_left', rsuffix='_right')

df
```

```python
df = df3.set_index('ID').join(df, how='left', lsuffix='_left', rsuffix='_right')
pd.set_option('display.max_columns', None)

df
```

```python
df = df.rename(columns = {'GDP_per_capita_USD_right' : 'GDP'})

df
```

```python
df = df[['Country', 'Year', 'Prevalence - Schizophrenia - Sex: Both - Age: Age-standardized (Percent)', 
         'Prevalence - Bipolar disorder - Sex: Both - Age: Age-standardized (Percent)', 
         'Prevalence - Eating disorders - Sex: Both - Age: Age-standardized (Percent)', 
         'Prevalence - Anxiety disorders - Sex: Both - Age: Age-standardized (Percent)', 
         'Prevalence - Drug use disorders - Sex: Both - Age: Age-standardized (Percent)', 
         'Prevalence - Depressive disorders - Sex: Both - Age: Age-standardized (Percent)', 
         'Prevalence - Alcohol use disorders - Sex: Both - Age: Age-standardized (Percent)', 
         'Total', 
         'population', 
         'GDP', 
         'Unemployment']]

df
```

```python
import matplotlib.font_manager as fm
```

```python
path = "MADEAvenuePERSONALUSE-Regular.otf"
path2 = "Philosopher-Regular.ttf"
path3 = "Quicksand_Bold.otf"
path4 = "GrebackGrotesquePERSONAL-Regular.otf"



font1 = fm.FontProperties(fname = path, size = 25)
font2 = fm.FontProperties(fname = path2, size = 13)
font3 = fm.FontProperties(fname = path3, size = 9)
font4 = fm.FontProperties(fname = path, size = 22)
font5 = fm.FontProperties(fname = path2, size = 12)
font6 = fm.FontProperties(fname = path2, size = 18)
```

```python
plt.figure(figsize = (14,4))

plt.bar(india['Year'], india['Unemployment'], color = '#235347', label = 'India')
plt.bar(indo['Year'], indo['Unemployment'], color = '#4f0205', label = 'Indonesia', bottom = india['Unemployment'])

plt.title('Tingkat Pengangguran di India dan Indonesia Tahun 1990 - 2019', fontproperties = font1)
plt.xlabel('Year', fontproperties = font2)
plt.ylabel('Unemployment(%)', fontproperties = font2)


plt.xticks(np.arange(1990, 2020, 2), fontproperties = font3)
plt.yticks(fontproperties = font3)

plt.legend(prop = font5)

plt.show()
```

```python
sns.jointplot(data = indo, y = indo['Total'], x = indo['Unemployment'], kind = 'reg', color = '#313f5f')

plt.title('Korelasi Antara Mental Health Disorder \ndengan Tingkat Pengangguran di Indonesia\n Tahun 1990 - 2019', pad = 80, fontproperties = font4)
plt.xlabel('Unemployment (%))', fontproperties = font2)
plt.ylabel('Mental Health Disorder (%)', fontproperties = font2)

plt.xticks(fontproperties = font3)
plt.yticks(fontproperties = font3)

plt.grid(linestyle = 'dotted')

plt.show()
```

```python
sns.jointplot(data = india, y = india['Total'], x = india['Unemployment'], kind = 'reg', color = '#6e6456')

plt.title('Korelasi Antara Mental Health Disorder \ndengan Tingkat Pengangguran di India\n Tahun 1990 - 2019', pad = 80, fontproperties = font4)
plt.xlabel('Unemployment (%))', fontproperties = font2)
plt.ylabel('Mental Health Disorder (%)', fontproperties = font2)

plt.xticks(fontproperties = font3)
plt.yticks(fontproperties = font3)

plt.grid(linestyle = 'dotted')

plt.show()
```
