```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
file = "prevalence-by-mental-and-substance-use-disorder.csv"

df = pd.read_csv(file)

df
```

```python
df = df.loc[(df['Entity'] == 'Indonesia') |
       (df['Entity'] == 'India')]

df
```

```python
df = df.drop([0,30], axis = 0)

df
```

```python
df['Total'] = df.iloc[:, [3, 4, 5, 6, 7, 8, 9]].sum(axis = 1)
```

```python
file2 = "populasi.csv"

df2 = pd.read_csv(file2, sep = ';')

df2
```

```python
populasi = df2['population']
```

```python
merge = pd.merge(df,populasi, on = populasi, how = 'left')
pd.set_option('display.max_rows', None)

merge
```

```python
df = merge.drop(columns = ['key_0'])

df
```

```python
file3 = "gdp.csv"

df3 = pd.read_csv(file3, sep = ';')

df3
```

```python
gdp = df3['GDP_per_capita_USD']

gdp
```

```python
merge = pd.merge(df,gdp, on = gdp, how = 'left')
pd.set_option('display.max_rows', None)

merge
```

```python
df = merge.drop(columns = ['key_0'])

df
```

```python
df = df.rename(columns = {'GDP_per_capita_USD' : 'GDP'})

df
```

```python
file4 = "unemployment.csv"

df4 = pd.read_csv(file4, sep = ';')

df4
```

```python
unemployment = df4['Unemployment']

unemployment
```

```python
merge = pd.merge(df,unemployment, on = unemployment, how = 'left')
pd.set_option('display.max_rows', None)

merge
```

```python
india = df.loc[df['Entity'] == 'India']

india
```

```python
indo = df.loc[df['Entity'] == 'Indonesia']

indo
```

```python
total_indo = indo['Total']

total_indo
```

```python
total_india = india['Total']

total_india
```

```python
gdp_indo = indo['GDP']

gdp_indo
```

```python
gdp_india = india['GDP']

gdp_india
```

```python
import matplotlib.font_manager as fm
```

```python
path = "MADEAvenuePERSONALUSE-Regular.otf"
path2 = "Philosopher-Regular.ttf"
path3 = "Quicksand_Bold.otf"
path4 = "GrebackGrotesquePERSONAL-Regular.otf"



font1 = fm.FontProperties(fname = path, size = 25)
font2 = fm.FontProperties(fname = path2, size = 13)
font3 = fm.FontProperties(fname = path3, size = 9)
font4 = fm.FontProperties(fname = path, size = 22)
font5 = fm.FontProperties(fname = path2, size = 12)
font6 = fm.FontProperties(fname = path2, size = 18)
```

```python
plt.figure(figsize = (14,4))

plt.bar(india['Year'], india['GDP'], color = '#dab7be', label = 'India')
plt.bar(indo['Year'], indo['GDP'], color = '#560c25', label = 'Indonesia', bottom = india['GDP'])

plt.title('Perkembangan GDP Per Kapita di India dan Indonesia Tahun 1990 - 2019', fontproperties = font1)
plt.xlabel('Year', fontproperties = font2)
plt.ylabel('GDP Per Capita (USD)', fontproperties = font2)


plt.xticks(np.arange(1990, 2020, 2), fontproperties = font3)
plt.yticks(fontproperties = font3)

plt.legend(prop = font5)

plt.show()
```

```python
sns.jointplot(data = indo, y = indo['Total'], x = indo['GDP'], kind = 'reg', color = '#66523a')

plt.title('Korelasi Antara Mental Health Disorder \ndengan GDP Per Kapita di Indonesia\n Tahun 1990 - 2019', pad = 80, fontproperties = font4)
plt.xlabel('GDP Per Capita (USD)', fontproperties = font2)
plt.ylabel('Mental Health Disorder (%)', fontproperties = font2)

plt.xticks(fontproperties = font3)
plt.yticks(fontproperties = font3)

plt.grid(linestyle = 'dotted')

plt.show()
```

```python
sns.jointplot(data = india, y = india['Total'], x = india['GDP'], kind = 'reg', color = '#344c11')

plt.title('Korelasi Antara Mental Health Disorder \ndengan GDP Per Kapita di India\n Tahun 1990 - 2019', pad = 80, fontproperties = font4)
plt.xlabel('GDP Per Capita (USD)', fontproperties = font2)
plt.ylabel('Mental Health Disorder (%)', fontproperties = font2)

plt.xticks(fontproperties = font3)
plt.yticks(fontproperties = font3)

plt.grid(linestyle = 'dotted')

plt.show()
```
